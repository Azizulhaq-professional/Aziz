- Client → Request → Server → Response

| Method   | Use                |
|----------|--------------------|
| `GET`    | Retrieve data      |
| `POST`   | Send data          |
| `PUT`    | Update             |
| `DELETE` | Remove             |

### Step 4 — FULL PROFESSIONAL app.py (WITH COMMENTS)

Copy this complete code into app.py.
```
# ==============================
# IMPORT LIBRARIES
# ==============================

from flask import Flask, request, jsonify
import joblib
import numpy as np

# ==============================
# CREATE FLASK APP
# ==============================

app = Flask(__name__)

# ==============================
# LOAD TRAINED MODEL
# ==============================

# Load ML model from model folder
model = joblib.load("model/churn_model.pkl")

# ==============================
# ROOT ROUTE (Health Check)
# ==============================

@app.route("/")
def home():
    """
    This route is used to check if API is alive.
    When user opens the website, they see this message.
    """
    return jsonify({
        "message": "Churn Prediction API is running 🚀"
    })

# ==============================
# GET API ENDPOINT
# ==============================

@app.route("/api", methods=["GET"])
def api_status():
    """
    Professional API health check endpoint.
    Used by developers & cloud platforms.
    """
    return jsonify({
        "status": "success",
        "message": "API is working perfectly"
    })

# ==============================
# POST API ENDPOINT (PREDICTION)
# ==============================

@app.route("/predict", methods=["POST"])
def predict():
    """
    This endpoint receives JSON data,
    sends it to ML model,
    and returns prediction.
    """

    # ---- Step 1: Get JSON from request ----
    data = request.get_json()

    # ---- Step 2: Validate input ----
    # If user sends empty request → show error
    if not data:
        return jsonify({
            "status": "error",
            "message": "No JSON data received"
        }), 400

    # Check if 'features' key exists
    if "features" not in data:
        return jsonify({
            "status": "error",
            "message": "Missing 'features' key in JSON"
        }), 400

    try:
        # ---- Step 3: Convert JSON → NumPy array ----
        features = np.array(data["features"]).reshape(1, -1)

        # ---- Step 4: Model Prediction ----
        prediction = model.predict(features)

        # ---- Step 5: Return Success Response ----
        return jsonify({
            "status": "success",
            "prediction": int(prediction[0])
        })

    except Exception as e:
        # ---- Step 6: Error Handling ----
        # If model crashes → send error message
        return jsonify({
            "status": "error",
            "message": str(e)
        }), 500


# ==============================
# RUN APP (LOCAL SERVER)
# ==============================

if __name__ == "__main__":
    app.run(debug=True)

```
