### Topic: Cloud Deployment (Make Your ML App LIVE)
### Learning Objectives

By the end of today, you will:

- Deploy your ML app to the cloud
- Understand hosting platforms
- Make your API accessible via a public URL
- Prepare a portfolio-ready project

### Step 1: What is Deployment?

👉 Deployment = Hosting your app on the internet

Instead of:
```
http://127.0.0.1:5000
```
You get:
```
https://your-app-name.onrender.com
```

### Step 2: Deployment Platforms

You can use:

- Render (Recommended for beginners)
- Railway
- Amazon Web Services (Advanced)

### 📁 Step 3: Prepare Project Files

Your project should include:
```
ML-App/
│
├── app.py
├── requirements.txt
├── Procfile
├── model/
│   └── churn_model.pkl
```

### Step 4: Create requirements.txt
```
pip freeze > requirements.txt
```

### Step 5: Create Procfile
```
web: python app.py
```

### Step 6: Modify Flask App
```
if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000)
```

### Step 7: Push Code to GitHub

👉 Use your repo:
- Upload project files
- Commit changes

### Step 8: Deploy on Render
Steps:
1. Go to Render
2. Create account
3. Click New Web Service
4. Connect GitHub repo
5. Deploy

### Step 9: Access Your Live App

👉 After deployment:
```
https://your-app-name.onrender.com
```

### Step 10: Test Live API

Using

Postman
- Send POST request
- Get prediction from live server

### Step 11: What You Achieved

- Built ML model

👉 You now have a complete ML product

### Step 12: Common Issues
- App not starting → check logs
- Missing libraries → update requirements.txt
- Model not loading → check file path
