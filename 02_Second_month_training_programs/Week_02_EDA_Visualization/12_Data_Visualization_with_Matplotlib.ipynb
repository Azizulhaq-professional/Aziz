### Step 0 — Import Libraries (Run First in Colab)

```python
# Matplotlib is used for plotting graphs
import matplotlib.pyplot as plt

# Pandas is used for handling datasets
import pandas as pd
```

### Step 1 — Line Plot (Trend Analysis)

👉 Shows how values change over time.

```python
# Create sample data
days = [1, 2, 3, 4, 5]
sales = [200, 250, 300, 280, 350]

# Create line plot
plt.plot(days, sales)

# Add title and labels
plt.title("Sales Over Days")
plt.xlabel("Days")
plt.ylabel("Sales")

# Show grid for better readability
plt.grid()

# Display the plot
plt.show()
```

What you learned:
- Line plots show trends
- Used in stock prices, growth, performance
###📊 Step 2 — Bar Chart (Category Comparison)

👉 Compare categories like products, countries, departments.

```python
# Create data
products = ["A", "B", "C", "D"]
sales = [50, 80, 30, 90]

# Create bar chart
plt.bar(products, sales)

# Add title and labels
plt.title("Product Sales Comparison")
plt.xlabel("Products")
plt.ylabel("Sales")

plt.show()
```

What you learned:
Bar chart compares categories
### Step 3 — Histogram (Distribution)

👉 Shows how frequently values occur.

```python
# Continuous data example
data = [10,20,20,30,30,30,40,40,50]

# Create histogram
plt.hist(data, bins=5)

# Add labels
plt.title("Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

What you learned:
- Histogram = frequency distribution
### Step 4 — Scatter Plot (Relationship)

👉 Shows correlation between two variables.

```python
# Example data
x = [1,2,3,4,5]
y = [2,4,5,4,5]

# Create scatter plot
plt.scatter(x, y)

# Add labels
plt.title("Relationship Between X and Y")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()
```

What you learned:
- Scatter plot finds relationship/correlation
### Step 5 — Plot Customization (INTERVIEW FAVORITE )

👉 Makes graphs professional.

```python
# Check  data lengths
print(len(days), len(sales))
```

```python
# New Data with five entries
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
sales = [100, 120, 90, 110, 95]  # now 5 values

# Styled line plot
plt.plot(days, sales,
         color='red',        # line color
         linestyle='--',     # dashed line
         marker='o')         # circle markers

plt.title("Styled Sales Graph")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.grid(True)

plt.show()
```

Must remember for interviews:
- color
- linestyle
- marker
- grid

### Task 1 — Line Plot of Revenue
