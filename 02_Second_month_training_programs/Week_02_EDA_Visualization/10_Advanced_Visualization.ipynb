### Project: Sales Data Analysis System

### STEP 1 — Import Libraries (Always run first)

```python
# Data analysis library
import pandas as pd

# Visualization library
import matplotlib.pyplot as plt
```

### STEP 2 — Create Sales Dataset

```python
# Create sales dataset (simulating real company sales data)
data = {
    "Product": ["Laptop", "Phone", "Tablet", "Laptop", "Phone", "Tablet"],
    "Region": ["North", "South", "East", "West", "North", "South"],
    "Sales": [50000, 30000, 20000, 60000, 35000, 25000],
    "Quantity": [5, 10, 8, 6, 12, 9]
}

# Convert dictionary into DataFrame
df = pd.DataFrame(data)

# Display dataset
df
```

### STEP 3 — Quick Data Overview (EDA)

```python
# Dataset information (data types, rows, missing values)
df.info()
```

```python
# Statistical summary (mean, min, max, etc.)
df.describe()
```

### STEP 4 — Total Sales by Product

```python
# Group by product and calculate total sales
product_sales = df.groupby("Product")["Sales"].sum()

product_sales
```

Shows which product generates most revenue.

### STEP 5 — Total Sales by Region

```python
# Group by region and calculate total sales
region_sales = df.groupby("Region")["Sales"].sum()

region_sales
```

### STEP 6 — Visualization
#### Bar Chart — Product Sales


```python
# Plot bar chart for product sales
product_sales.plot(kind="bar")

plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")

plt.show()
```

#### Pie Chart — Region Sales Distribution

```python
# Pie chart showing percentage distribution by region
region_sales.plot(kind="pie", autopct="%1.1f%%")

plt.title("Sales Distribution by Region")
plt.ylabel("")  # Remove extra label

plt.show()
```

#### Line Plot — Sales Trend

```python
# Line plot showing sales trend across entries
df["Sales"].plot(kind="line")

plt.title("Sales Trend")
plt.xlabel("Index")
plt.ylabel("Sales")

plt.show()
```

### STEP 7 — Business Insights (Answer Questions)

Run these analysis cells:

```python
# Which product sells the most?
product_sales.sort_values(ascending=False)
```

```python
# Which region performs best?
region_sales.sort_values(ascending=False)
```

Based on output, you can say:
- Highest selling product = Laptop
- Best performing region = North / West (depending on totals)
### STEP 8 — Create New Feature (Feature Engineering)

👉 Revenue per item = Sales ÷ Quantity

```python
# Create new column
df["Revenue_per_Item"] = df["Sales"] / df["Quantity"]

# Display updated dataset
df
```
