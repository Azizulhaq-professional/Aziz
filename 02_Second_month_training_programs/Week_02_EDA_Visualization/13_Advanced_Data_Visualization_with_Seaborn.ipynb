### Step 0 — Install & Import Libraries (Run first)

```python
# Install seaborn in Colab (already installed usually, but safe to run)
!pip install seaborn

# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

### Step 1 — Load Built-in Dataset

Seaborn has free datasets we can use for practice.

```python
# Load restaurant tips dataset
df = sns.load_dataset("tips")

# Show first 5 rows
df.head()
```

Dataset Columns:
- total_bill → bill amount
- tip → tip given
- sex → male/female
- day → day of week
- time → lunch/dinner

Step 2 — Seaborn Scatter Plot

👉 Relationship between bill and tip.

```python
# Create scatter plot
sns.scatterplot(x="total_bill", y="tip", data=df)

# Add title
plt.title("Total Bill vs Tip")

plt.show()
```

Insight: Higher bill → Higher tip (positive relationship)

### Step 3 — Seaborn Bar Plot (Auto Average)

Seaborn automatically calculates average values

```python
# Average bill per day
sns.barplot(x="day", y="total_bill", data=df)

plt.title("Average Bill per Day")
plt.show()
```

Shows average spending per day.

### Step 4 — Box Plot (Very Important )

Used heavily in interviews.

```python
# Boxplot shows distribution & outliers
sns.boxplot(x="day", y="total_bill", data=df)

plt.title("Boxplot of Bills by Day")
plt.show()
```

Boxplot shows:
- Median (middle line)
- Data spread
- Outliers (dots)

👉 Queestion answer: Detect outliers + distribution

### Step 5 — Correlation Heatmap (MOST IMPORTANT)

```python
# Calculate correlation between numeric columns
corr = df.corr(numeric_only=True)

# Draw heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

How to read:
- +1 → Strong positive
- -1 → Strong negative
- 0 → No relation

👉 Used in feature selection for ML

### Step 6 — Pairplot (Full EDA in One Line )

```python
# Visualize relationship between ALL variables
sns.pairplot(df)

plt.show()
```

This is a complete mini EDA in one command.

### Step 7 — Distribution Plot + KDE

```python
# Distribution of total bill
sns.histplot(df["total_bill"], kde=True)

plt.title("Total Bill Distribution")
plt.show()
```

KDE = Smooth curve of distribution.
### Step 8 — Styling Graphs Professionally

```python
# Change global style
sns.set_style("darkgrid")

sns.barplot(x="day", y="total_bill", data=df)
plt.title("Styled Bar Plot")

plt.show()
```

Available styles:
- darkgrid
- whitegrid
- white
- ticks
### Step 9 — REAL PRACTICE TASK (Iris Dataset)

Load dataset:

```python
# Load iris dataset
iris = sns.load_dataset("iris")
iris.head()
```

### Task 1 — Scatter Plot

Sepal Length vs Sepal Width

```python
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris)

plt.title("Sepal Length vs Width")
plt.show()
```

### Task 2 — Boxplot

Species vs Petal Length

```python
sns.boxplot(x="species", y="petal_length", data=iris)

plt.title("Petal Length by Species")
plt.show()
```

### Task 3 — Heatmap (Correlation)

```python
# Correlation matrix
corr = iris.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Iris Correlation Heatmap")

plt.show()
```
