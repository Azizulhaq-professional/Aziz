## Topic: Exploratory Data Analysis (EDA)
### Why EDA is VERY IMPORTANT

👉 Before building models or reports, you must:
- Understand the data
- Find patterns
- Detect problems
- Discover insights

👉 This process is called:
- Exploratory Data Analysis (EDA)

### STEP 1 — Import Pandas & Create Dataset

```python
# Import pandas
import pandas as pd

# Create employee dataset
data = {
    "Name": ["Ali", "Ahmed", "Sara", "Zain", "Ayesha", "Bilal"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT"],
    "Salary": [50000, 60000, 45000, 70000, 80000, 55000],
    "Age": [25, 30, 22, 28, 35, 27]
}

# Convert dictionary → DataFrame
df = pd.DataFrame(data)

# Display dataset
df
```

### STEP 2 — Basic Data Overview

These commands help you understand structure of data.

```python
# Show first 5 rows
df.head()
```

```python
# Show last 5 rows
df.tail()
```

```python
# Number of rows and columns
df.shape
```

```python
# Column names
df.columns
```

```python
# Complete dataset info
df.info()
```

👉 From this you learn:
- How many rows?
- Which columns?
- Data types?
- Missing values?

### STEP 3 — Statistical Summary

```python
# Summary statistics for numeric columns
df.describe()
```

You get:
- Mean salary
- Min / Max salary
- Average age
- Standard deviation
### STEP 4 — Check Unique Values (Categorical Analysis)
#### Unique departments

```python
df["Department"].unique()
```

#### Count employees in each department

```python
df["Department"].value_counts()
```

👉 Helps answer:

Which department has more employees?

### STEP 5 — Analyze Relationships
Average salary per department

```python
df.groupby("Department")["Salary"].mean()
```

#### Highest salary in company

```python
df["Salary"].max()
```

### Lowest salary in company

```python
df["Salary"].min()
```

### STEP 6 — Find Business Insights
#### Which department pays the most?

```python
df.groupby("Department")["Salary"].mean().sort_values(ascending=False)
```

#### Youngest employee

```python
df.sort_values("Age").head(1)
```

#### Oldest employee

```python
df.sort_values("Age", ascending=False).head(1)
```

### Employees per department

```python
df["Department"].value_counts()
```

### STEP 7 — Correlation (Numbers Relationship)

```python
# Correlation between numeric columns
df.corr(numeric_only=True)
```

👉 Shows relationship between:
- Age vs Salary

Value meaning:
- Close to 1 → strong positive relation
- Close to -1 → strong negative relation
- Close to 0 → no relation

### STEP 8 — Detect Outliers

Find unusually high salaries.

```python
# Sort by salary (highest first)
df.sort_values("Salary", ascending=False)
```

This helps spot:
- Highest paid employee
- Possible outliers
