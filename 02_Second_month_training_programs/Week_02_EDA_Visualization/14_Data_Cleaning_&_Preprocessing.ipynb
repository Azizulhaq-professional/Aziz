### Step 0 — Import Libraries

```python
# Pandas → data handling
import pandas as pd

# Numpy → numerical operations
import numpy as np
```

### Step 1 — Create Practice Dataset (Given Task)

We will create the dataset.

```python
# Create messy dataset
data = {
    "age": [25, 30, None, 22, 30],
    "salary": [50000, 60000, 700000, 52000, None],
    "city": ["Lahore", "Karachi", None, "Islamabad", "Lahore"]
}

df = pd.DataFrame(data)

# Show dataset
df
```

Notice problems
- Missing values
- Possible outlier (700000 salary)
### Step 2 — Basic Data Inspection

Always inspect before cleaning.

```python
# Info about dataset
df.info()

# Statistical summary
df.describe()

# Count missing values
df.isnull().sum()
```

### Step 3 — Handle Missing Values
#### 👉 Fill numerical column with MEAN

```python
# Fill missing values in age with average age
df["age"].fillna(df["age"].mean(), inplace=True)

# Fill missing salary with average salary
df["salary"].fillna(df["salary"].mean(), inplace=True)
```

### Fill categorical column with MODE / label

```python
# Replace missing city with "Unknown"
df["city"].fillna("Unknown", inplace=True)

df
```

### Step 4 — Remove Duplicates

```python
# Remove duplicate rows (if any)
df = df.drop_duplicates()

df
```

### Step 5 — Fix Data Types

After filling missing values, convert age to integer.

```python
# Convert age from float → int
df["age"] = df["age"].astype(int)

df.info()
```

### Step 6 — Detect Outliers (IQR Method)

We check outliers in salary column.

```python
# Calculate quartiles
Q1 = df["salary"].quantile(0.25)
Q3 = df["salary"].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define lower & upper limits
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Show outliers
outliers = df[(df["salary"] < lower) | (df["salary"] > upper)]
outliers
```

You will see 700000 as outlier

### Step 7 — Remove Outliers

```python
# Keep only normal salary values
df = df[(df["salary"] >= lower) & (df["salary"] <= upper)]

df
```

Now dataset is clean

### Step 8 — Convert Categorical Data → Numbers

ML models don’t understand text.

```python
# Convert city column to numbers using one-hot encoding
df = pd.get_dummies(df, columns=["city"], drop_first=True)

df
```

### Step 9 — Feature Scaling (Very Important)

Bring salary to same scale.

```python
from sklearn.preprocessing import StandardScaler

# Create scaler object
scaler = StandardScaler()

# Scale salary column
df["salary"] = scaler.fit_transform(df[["salary"]])

df
```

### Step 10 — Final Clean Data Check

```python
# Check missing values again
df.isnull().sum()

# Final dataset info
df.info()
```

Dataset is now ML ready
