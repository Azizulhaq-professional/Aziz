## Topic: Data Visualization (Matplotlib Basics)
#### Why Visualization is IMPORTANT

👉 Raw data is hard to understand

👉 Graphs make insights clear and powerful

Used in:
- Dashboards
- Business reports
- Presentations
- Interviews

### STEP 1 — Import Libraries

```python
# Pandas → data handling
import pandas as pd

# Matplotlib → plotting graphs
import matplotlib.pyplot as plt
```

### STEP 2 — Create Dataset

```python
# Create employee dataset
data = {
    "Name": ["Ali", "Ahmed", "Sara", "Zain", "Ayesha", "Bilal"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT"],
    "Salary": [50000, 60000, 45000, 70000, 80000, 55000],
    "Age": [25, 30, 22, 28, 35, 27]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display dataset
df
```

### STEP 3 — Line Plot (Salary Trend)

👉 Used to show trends over time/index

```python
# Plot salary values as a line graph
plt.plot(df["Salary"])

# Add title and labels
plt.title("Salary Trend")
plt.xlabel("Index")
plt.ylabel("Salary")

# Display graph
plt.show()
```

Shows how salary values change across employees.

### STEP 4 — Bar Chart (Category Comparison)

👉 Compare average salary by department

```python
# Group by department and calculate average salary
df.groupby("Department")["Salary"].mean().plot(kind="bar")

# Add labels and title
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Salary")

plt.show()
```

Very common in business dashboards.

### STEP 5 — Histogram (Distribution)

👉 Shows salary distribution

```python
# Histogram of salary values
plt.hist(df["Salary"])

plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")

plt.show()
```

Helps detect:
- Most common salary range
- Outliers
### STEP 6 — Scatter Plot (Relationship)

👉 Shows relationship between Age and Salary

```python
# Scatter plot Age vs Salary
plt.scatter(df["Age"], df["Salary"])

plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")

plt.show()
```

Used to check correlations.

###  STEP 7 — Professional Looking Chart

👉 Improve chart size + grid (important for reports)

```python
# Set figure size (width, height)
plt.figure(figsize=(8,5))

# Plot average salary per department again
df.groupby("Department")["Salary"].mean().plot(kind="bar")

# Add professional labels
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Salary")

# Add grid for readability
plt.grid()

plt.show()
```

# WHEN TO USE WHICH CHART

| Chart | Purpose |
|-------|---------|
| Line | Trends |
| Bar | Category comparison |
| Histogram | Distribution |
| Scatter | Relationships |
