### Topic: Handling Missing Data (Data Cleaning)
Why This Topic is CRITICAL

👉 In real datasets:
- Data is never clean
- Missing values are everywhere

👉 If you don’t handle them:
- Wrong analysis
- Bad ML models
-  Poor decisions

### STEP 1 — Create Dataset With Missing Values

```python
# Import libraries
import pandas as pd
import numpy as np   # Used to create missing values (NaN)

# Create dataset with missing values
data = {
    "Name": ["Ali", "Ahmed", "Sara", "Zain", "Ayesha"],
    "Age": [25, np.nan, 22, 28, np.nan],      # Missing ages
    "Salary": [50000, 60000, np.nan, 70000, 80000]  # Missing salary
}

# Convert into DataFrame
df = pd.DataFrame(data)

# Display dataset
df
```

np.nan = Missing value in Pandas.

###  STEP 2 — Detect Missing Values
#### Check where values are missing

```python
# Shows True/False for each cell
df.isnull()
```

True = missing value
False = value exists

#### Count missing values in each column

```python
# Count total missing values column-wise
df.isnull().sum()
```

You will see something like:
- Age → 2 missing
- Salary → 1 missing

### STEP 3 — Remove Missing Values

 Usually NOT best practice, but important to learn.
###  Remove rows that contain ANY missing value

```python
# Drop rows that have at least one NaN
df.dropna()
```

👉 Whole row removed if even one value is missing

####  Remove rows where Age is missing only

```python
# Drop rows where Age is NaN
df.dropna(subset=["Age"])
```

👉 Only checks Age column.

### STEP 4 — Fill Missing Values (Best Practice)

 In real projects we fill, not delete.

#### Fill Age with Mean

```python
# Replace missing Age values with average age
df["Age"].fillna(df["Age"].mean(), inplace=True)

df
```

Why mean?

Because Age is numeric data.

###  Fill Salary with Median

```python
# Replace missing Salary with median salary
df["Salary"].fillna(df["Salary"].median(), inplace=True)

df
```

Why median?

Median works better when outliers exist.

#### Fill With Constant Value (Example)

Sometimes we fill with fixed value.

```python
# Example: Replace any remaining missing Age with 0
df["Age"].fillna(0, inplace=True)

df
```

(After previous steps, this may not change anything — just for learning)

# 📊 WHEN TO USE WHAT?

| Situation |  Recommended Method |
|:----------|:----------------------|
| **Numeric data** | Mean / Median |
| **Outliers present** | Median |
| **Categorical data** | Mode |
| **Important data missing** |  Do NOT drop rows |

### STEP 5 — Check Again (Final Verification)

```python
# Final check for missing values
df.isnull().sum()
```

Output should be 0 missing values.
