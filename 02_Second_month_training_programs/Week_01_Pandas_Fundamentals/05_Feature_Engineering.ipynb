## Topic: Feature Engineering (Creating New Columns)
### Why This is IMPORTANT

You need to:
- Create new features
- Transform data
- Make it useful for analysis & ML

👉 This is called:

Feature Engineering (very important in AI & Data Science)

### STEP 1 — Import Pandas

````python
# Import pandas library
import pandas as pd
````

### STEP 2 — Create Dataset With Problems

Dataset has:
- Duplicate rows
- Salary stored as text

````python
# Create dataset
data = {
    "Name": ["Ali", "Ahmed", "Sara", "Ali", "Zain"],
    "Age": [25, 30, 22, 25, 28],
    "Salary": ["50000", "60000", "45000", "50000", "70000"]  # Salary as string
}

# Convert into DataFrame
df = pd.DataFrame(data)

# Show dataset
df
````

### STEP 3 — Detect Duplicate Rows

````python
# Check which rows are duplicates (True = duplicate)
df.duplicated()
````

####  Count total duplicate rows

````python
# Count number of duplicate rows
df.duplicated().sum()
````

### STEP 4 — Remove Duplicate Rows

#### VERY IMPORTANT CONCEPT
 Wrong
 ```
 df.drop_duplicates()
 ```
 Correct

We must assign back to dataframe.
```
# Remove complete duplicate rows
df = df.drop_duplicates()

df
```

````python
# Remove complete duplicate rows
df = df.drop_duplicates()

df
````

### STEP 5 — Remove Duplicates Based on Name Only

Sometimes we want one record per person.

````python
# Keep first occurrence of each Name only
df = df.drop_duplicates(subset=["Name"])

df
````

Now each employee appears once

### STEP 6 — Check Data Types

````python
# Check column data types
df.dtypes
````

You will see:

Salary → object  (string)

We need numeric type for calculations.
### STEP 7 — Convert Salary to Integer

````python
# Convert Salary from string to integer
df["Salary"] = df["Salary"].astype(int)

# Verify conversion
df.dtypes
````

Now Salary is numeric

### STEP 8 — Rename Columns to Lowercase
- Best practice in data science:
- Column names should be lowercase.

````python
# Rename columns
df.columns = ["name", "age", "salary"]

df
````

### STEP 9 — Clean Text Data (Standardization)

Why?

"Ali" and "ali" should be same.

````python
# Convert names to lowercase
df["name"] = df["name"].str.lower()

df
````

### STEP 10 — Final Clean Dataset

````python
# Final cleaned dataset
print(df)
````
