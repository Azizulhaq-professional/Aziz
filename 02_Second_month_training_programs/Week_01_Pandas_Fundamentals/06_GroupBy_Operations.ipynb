## Topic: GroupBy Operations (Core Data Analysis Skill)
###  Why GroupBy is VERY IMPORTANT

👉 This is used in:
- Business reports
- Dashboards
- Data Analyst jobs

👉 It helps you answer questions like:
- Total sales by category
- Average salary by department
- Count of employees by group

### STEP 1 — Import Pandas & Create Dataset

```python
# Import pandas
import pandas as pd

# Create employee dataset
data = {
    "Name": ["Ali", "Ahmed", "Sara", "Zain", "Ayesha", "Bilal"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT"],
    "Salary": [50000, 60000, 45000, 70000, 80000, 55000]
}

# Convert dictionary into DataFrame
df = pd.DataFrame(data)

# Show dataset
df
```

### STEP 2 — Basic GroupBy

```python
# Create groups based on Department
df.groupby("Department")
```

👉 This only creates groups internally.

To see results, we must apply aggregation (mean, sum, count).

### STEP 3 — Aggregation Functions
#### Average Salary by Department

```python
# Mean salary for each department
df.groupby("Department")["Salary"].mean()
```

This answers:

👉 Which department pays average salary highest?

####  Total Salary by Department

```python
# Total salary spent per department
df.groupby("Department")["Salary"].sum()
```

This answers:

👉 How much company spends per department

#### Count Employees per Department

```python
# Count number of employees in each department
df.groupby("Department")["Name"].count()
```

This answers:

👉 How many employees in each department

#### STEP 4 — Multiple Aggregations Together

```python
# Apply multiple statistics at once
df.groupby("Department")["Salary"].agg(["mean", "sum", "max", "min"])
```

You get:
- Average salary
- Total salary
- Highest salary
- Lowest salary

This is very common in dashboards

### STEP 5 — Reset Index (Clean Table Output)

GroupBy makes Department an index.

We convert it back to normal column.

```python
# Convert result back to normal table format
df.groupby("Department")["Salary"].mean().reset_index()
```

Now output looks like a proper table

### STEP 6 — GroupBy Multiple Columns

```python
# Group by Department AND Salary
df.groupby(["Department", "Salary"]).count()
```

This creates multi-level grouping.

Used in advanced reports.
### STEP 7 — Real-World Insight Example
####Find Highest Paying Department

```python
# Sort departments by average salary (highest first)
df.groupby("Department")["Salary"].mean().sort_values(ascending=False)
```

This answers real business question:

👉 Which department pays the most?
