## Topic: Merging & Joining DataFrames
#### Why This is VERY IMPORTANT

👉 In real-world projects:
- Data is stored in multiple tables/files
- You must combine them

👉 Just like SQL JOINs:
- Customer data + Orders
- Employees + Salaries

### STEP 1 — Import Pandas & Create Two Datasets

```python
# Import pandas
import pandas as pd

# Employees table (like Employees.csv)
employees = pd.DataFrame({
    "Emp_ID": [1, 2, 3, 4],
    "Name": ["Ali", "Ahmed", "Sara", "Zain"],
    "Department": ["IT", "HR", "Finance", "IT"]
})

# Salaries table (like Salaries.csv)
salaries = pd.DataFrame({
    "Emp_ID": [1, 2, 3, 5],
    "Salary": [50000, 60000, 70000, 80000]
})

# Display both datasets
employees, salaries
```

👉 Real world: data usually comes from different files/tables.

### STEP 2 — INNER JOIN (Common Records Only)

```python
# Merge where Emp_ID exists in BOTH tables
merged = pd.merge(employees, salaries, on="Emp_ID", how="inner")

merged
```

Explanation:
- Emp_ID 1,2,3 exist in both → kept
- Emp_ID 4 & 5 removed (not common)
### STEP 3 — LEFT JOIN (Keep All Employees)

```python
# Keep ALL employees, even if salary missing
left_join = pd.merge(employees, salaries, on="Emp_ID", how="left")

left_join
```

👉 Employee with Emp_ID=4 has no salary → NaN

This is VERY common in real datasets.

### STEP 4 — RIGHT JOIN (Keep All Salaries)

```python
# Keep ALL salary records
right_join = pd.merge(employees, salaries, on="Emp_ID", how="right")

right_join
```

👉 Emp_ID=5 appears even though employee data missing.

### STEP 5 — OUTER JOIN (Keep EVERYTHING)

```python
# Keep all rows from both tables
outer_join = pd.merge(employees, salaries, on="Emp_ID", how="outer")

outer_join
```

This shows complete combined dataset.

### STEP 6 — Handle Missing Values After Merge
- After LEFT JOIN we saw NaN.
- We often replace missing values.

```python
# Replace NaN with 0
left_join.fillna(0, inplace=True)

left_join
```

### STEP 7 — REAL WORLD ANALYSIS

👉 Total Salary by Department

##### First merge tables again:

```python
# Merge employees and salaries (inner join)
merged = pd.merge(employees, salaries, on="Emp_ID", how="inner")

merged
```

#### Now perform analysis:

```python
# Total salary spent per department
merged.groupby("Department")["Salary"].sum()
```

### Business question answered:

How much company spends per department
