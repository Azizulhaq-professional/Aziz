### STEP 1 — Import Pandas

Create a new Colab notebook and run:

```python
import pandas as pd
```

### STEP 2 — Create Your Own Dataset

We will create a Student dataset (easy to understand).

```python
data = {
    "Name": ["Ali", "Ahmed", "Sara", "Fatima", "Usman", "Ayesha", "Bilal", "Zain"],
    "Age": [21, 22, 20, 23, 24, 21, 22, 23],
    "City": ["Lahore", "Karachi", "Islamabad", "Lahore", "Karachi", "Islamabad", "Lahore", "Karachi"],
    "Marks": [85, 90, 78, 88, 92, 95, 80, 87]
}

df = pd.DataFrame(data)
df
```

 This creates a DataFrame (table).

### STEP 3 — Save Dataset as CSV (Important for Task)

Now we save it as a CSV file.

```python
df.to_csv("students.csv", index=False)
```

Now your dataset is stored in Colab storage.

### STEP 4 — Load CSV using Pandas

This matches your task requirement

```python
df = pd.read_csv("students.csv")
df
```
