### STEP 1 — Load Dataset in Colab

```python
# Import pandas library
import pandas as pd

# Create dataset (Employee data)
data = {
    "Name": ["Ali", "Ahmed", "Sara", "Zain", "Ayesha"],
    "Age": [25, 30, 22, 28, 35],
    "Salary": [50000, 60000, 45000, 70000, 80000]
}

# Convert dictionary into DataFrame (table format)
df = pd.DataFrame(data)

# Show dataset
df
```

### STEP 2 — Filtering Data (Conditions)
Example 1: Age greater than 25

```python
# Show employees whose Age is greater than 25
df[df["Age"] > 25]
```

What happens here?
- df["Age"] > 25 creates True/False values
- Pandas shows only rows where condition is True

#### Example 2: Salary greater than 60,000

```python
# Show employees earning more than 60k
df[df["Salary"] > 60000]
```

Real meaning: Find high earners

####  Example 3: Multiple Conditions (AND)

```python
# Employees older than 25 AND salary above 60k
df[(df["Age"] > 25) & (df["Salary"] > 60000)]
```

Important:
- Use & for AND
- Wrap each condition in brackets

#### Example 4: OR Condition

```python
# Employees older than 30 OR salary above 70k
df[(df["Age"] > 30) | (df["Salary"] > 70000)]
```

Use | for OR

### STEP 3 — Sorting Data
 Sort by Age (Ascending)

```python
# Sort employees by Age (young → old)
df.sort_values("Age")
```

Ascending = smallest → largest (default)

#### Sort by Salary (Descending)

```python
# Sort employees by Salary (highest → lowest)
df.sort_values("Salary", ascending=False)
```

ascending=False = biggest → smallest

### STEP 4 — Select Specific Columns After Filtering

```python
# Show only Name & Salary where Age > 25
df[df["Age"] > 25][["Name", "Salary"]]
```

Step breakdown:

1. Filter rows → Age > 25
2. Select columns → Name & Salary

This is very common in real projects
