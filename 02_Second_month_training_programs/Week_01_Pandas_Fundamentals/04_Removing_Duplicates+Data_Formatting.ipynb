## Topic: Removing Duplicates + Data Formatting
### Why This Matters

In real datasets:
- Same record appears multiple times
- Data types are incorrect

👉 If you don’t fix this:
- Wrong analysis
- Misleading insights

### STEP 1 — Create Dataset With Issues

```python
# Import pandas
import pandas as pd

# Dataset contains problems:
# - Duplicate rows
# - Salary stored as string (object) instead of number

data = {
    "Name": ["Ali", "Ahmed", "Sara", "Ali", "Zain"],
    "Age": [25, 30, 22, 25, 28],
    "Salary": ["50000", "60000", "45000", "50000", "70000"]  #  string values
}

df = pd.DataFrame(data)

# Display dataset
df
```

### STEP 2 — Check Duplicate Rows

```python
# Shows True/False for duplicate rows
df.duplicated()
```

True means row is duplicated.

#### Count total duplicates

```python
df.duplicated().sum()
```

This tells how many duplicate rows exist.

### STEP 3 — Remove Duplicate Rows

```python
# Remove complete duplicate rows
df = df.drop_duplicates()

df
```

Now identical rows are removed.

### STEP 4 — Remove Duplicates Based on Column

Sometimes we want one record per person.

```python
# Keep only first occurrence of each Name
df = df.drop_duplicates(subset=["Name"])

df
```

Now each name appears only once.

### STEP 5 — Check Data Types

```python
df.dtypes
```

You will notice:

| Column | Type |
|--------|------|
| Name | object |
| Age | int |
| Salary | object  (should be numeric) |

Salary should be numeric.

### STEP 6 — Convert Data Type

```python
# Convert Salary from string → integer
df["Salary"] = df["Salary"].astype(int)

# Verify again
df.dtypes
```

Now Salary becomes int64

### STEP 7 — Rename Columns (Standard Naming)

Best practice in data science → use lowercase column names.

```python
# Rename columns
df.columns = ["name", "age", "salary"]

df
```

### STEP 8 — Clean Text Data

Make names consistent (lowercase).

```python
# Convert names to lowercase
df["name"] = df["name"].str.lower()

df
```

Why?
Because "Ali" and "ali" should be treated same.

### STEP 9 — Final Clean Dataset

```python
# Final cleaned dataset
print(df)
```

You now have:
- No duplicates
- Correct data types
- Clean column names
- Consistent text
