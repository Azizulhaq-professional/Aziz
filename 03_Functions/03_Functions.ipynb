### 1. What is a Function?

👉 A function is a reusable block of code that performs a task.

🔹 Why use functions?
- Avoid repetition
- Make code clean
- Easy to debug
- Used in ALL real systems
### 2. Basic Function (No Input, No Return)

```python
def greet():
    print("Hello, welcome!")
```

```python
# Call function:
greet()
```

Explanation:
- def → defines function
- greet → function name
- () → parameters (empty here)
### 3. Function with Parameters (Input)

```python
def greet(name):
    print(f"Hello {name}")
```

```python
# Call:
greet("Aziz")
```

Explanation:
- name is input parameter
- Value is passed when calling
###  4. Function with Multiple Parameters

```python
def add(a, b):
    print(a + b)
```

```python
# Call:
add(5, 3)
```

### 5. Function with Return (VERY IMPORTANT)

```python
def add(a, b):
    return a + b
```

```python
# Call:
result = add(5, 3)
print(result)
```

Explanation:
- return sends value back
- Used in ML, APIs, backend
### 6. Function with Default Parameters

```python
def greet(name="User"):
    print(f"Hello {name}")
```

```python
# Call:
greet()        # Hello User
greet("Ali")   # Hello Ali
```

### 7. Function with List Input

```python
def print_numbers(numbers):
    for num in numbers:
        print(num)
```

```python
# Call:
print_numbers([10, 20, 30])
```

### 8. Function Returning Multiple Values

```python
def calculate(a, b):
    return a + b, a * b
```

```python
sum_val, mul_val = calculate(5, 3)
print(sum_val, mul_val)
```

### 9. Nested Function Call

```python
def square(x):
    return x * x

def calculate():
    result = square(5)
    print(result)

calculate()
```

### 10. Lambda Function (Short Function)

```python
add = lambda a, b: a + b
print(add(5, 3))
```

👉 Used in:

- Data Science
- Pandas
- ML pipelines
### 11. Function with Error Handling (Professional Level)

```python
def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Cannot divide by zero"
```

```python
# Call:
print(divide(10, 0))
```

### MAIN TASK — Average Function
Final Professional Version

```python
def calculate_average(numbers):
    if len(numbers) == 0:
        return 0
    return sum(numbers) / len(numbers)
```

```python
# Use:
numbers = [10, 20, 30, 40, 50]
avg = calculate_average(numbers)
print(f"Average: {avg}")
```

### FULL PROGRAM (INDUSTRY STYLE)

```python
def get_numbers():
    numbers = []

    while True:
        user_input = input("Enter number (or 'done' to stop): ")

        if user_input.lower() == "done":
            break

        try:
            num = float(user_input)
            numbers.append(num)
        except ValueError:
            print("Invalid input")

    return numbers


def calculate_average(numbers):
    if len(numbers) == 0:
        return 0
    return sum(numbers) / len(numbers)


def main():
    print("Average Calculator\n")

    numbers = get_numbers()
    avg = calculate_average(numbers)

    print(f"\nNumbers: {numbers}")
    print(f"Average: {avg}")


if __name__ == "__main__":
    main()
```

### PROJECT: Dynamic Average Calculator
- User can enter unlimited numbers
- Type "done" to stop
- Calculates:
  - Average
  - Sum
  - Maximum value

```python
def get_numbers():
    numbers = []

    print("Enter numbers (type 'done' to finish):")

    while True:
        user_input = input("Enter number: ")

        if user_input.lower() == "done":
            break

        try:
            num = float(user_input)
            numbers.append(num)
        except ValueError:
            print("Invalid input! Please enter a valid number.")

    return numbers


def calculate_sum(numbers):
    return sum(numbers)


def calculate_average(numbers):
    if len(numbers) == 0:
        return 0
    return sum(numbers) / len(numbers)


def find_max(numbers):
    if len(numbers) == 0:
        return None
    return max(numbers)


def display_results(numbers):
    print("\n--- Results ---")
    print(f"Numbers: {numbers}")
    print(f"Total Sum: {calculate_sum(numbers)}")
    print(f"Average: {calculate_average(numbers)}")

    max_value = find_max(numbers)
    if max_value is not None:
        print(f"Maximum Value: {max_value}")
    else:
        print("No numbers entered.")


def main():
    print("📊 Dynamic Number Analyzer\n")

    numbers = get_numbers()
    display_results(numbers)


if __name__ == "__main__":
    main()
```
