### PROJECT
### Project Name:

👉 Student API Simulator (Local JSON System)

### Features:
- Add student
- Save data in JSON file
- Read JSON file
- Display data

### FULL PROFESSIONAL CODE

```python
import json

# Function to add student
def add_student():
    name = input("Enter student name: ")

    while True:
        try:
            marks = float(input("Enter marks: "))
            break
        except ValueError:
            print("Invalid input")

    city = input("Enter city: ")

    return {
        "name": name,
        "marks": marks,
        "city": city
    }


# Function to save data to JSON file
def save_to_json(student):
    try:
        # Load existing data
        with open("students.json", "r") as file:
            data = json.load(file)
    except (FileNotFoundError, json.JSONDecodeError):
        data = []

    # Add new student
    data.append(student)

    # Save back to file
    with open("students.json", "w") as file:
        json.dump(data, file, indent=4)


# Function to read JSON file
def read_json():
    print("\n--- Student Data (JSON) ---")

    try:
        with open("students.json", "r") as file:
            data = json.load(file)

            for i, student in enumerate(data, start=1):
                print(f"\nStudent {i}:")
                for key, value in student.items():
                    print(f"{key.capitalize()}: {value}")

    except FileNotFoundError:
        print("No data found.")


# Main function
def main():
    print("Student API Simulator\n")

    while True:
        student = add_student()
        save_to_json(student)

        choice = input("Add another student? (yes/no): ").lower()
        if choice != "yes":
            break

    read_json()


if __name__ == "__main__":
    main()
```
