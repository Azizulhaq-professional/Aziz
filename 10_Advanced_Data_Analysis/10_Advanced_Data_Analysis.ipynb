### Step 1 — Create sales_data.csv inside Colab

```python
import pandas as pd

data = {
    "name": ["Laptop", "Mouse", "Keyboard", "Monitor", None, "Mouse", "Laptop", "Tablet", "Phone", "Headphones"],
    "price": [1200, 25, 45, None, 300, 25, 1200, 400, None, 80],
    "quantity": [1, 2, None, 1, 3, 2, 1, None, 5, 2],
    "city": ["Islamabad","Lahore","Karachi","Islamabad","Lahore","Lahore","Islamabad","Karachi","Islamabad","Lahore"]
}

df = pd.DataFrame(data)

# Save CSV file
df.to_csv("sales_data.csv", index=False)

df
```

### Step 2 — Now run YOUR cleaning code

Run this in the next cell:

```python
import pandas as pd

df = pd.read_csv("sales_data.csv")

# Clean data
df["price"].fillna(df["price"].mean(), inplace=True)
df["quantity"].fillna(1, inplace=True)
df = df.drop_duplicates()
df = df.dropna(subset=["name"])

# Create total column
df["total"] = df["price"] * df["quantity"]

df
```

### Step 3 — Save the CLEAN dataset (important)

```python
df.to_csv("sales_data_cleaned.csv", index=False)
```

### PROJECT (CLEAN + ANALYSIS)

```python
import pandas as pd

# Load data
df = pd.read_csv("sales_data.csv")

# Cleaning
df["price"].fillna(df["price"].mean(), inplace=True)
df["quantity"].fillna(1, inplace=True)
df = df.drop_duplicates()
df = df.dropna(subset=["name"])

# Feature engineering
df["total"] = df["price"] * df["quantity"]

# Sorting
print("\n Top Sales:")
print(df.sort_values(by="total", ascending=False).head(3))

# Product analysis
print("\n Product Sales:")
print(df.groupby("name")["total"].sum())

# City analysis
print("\n City Revenue:")
print(df.groupby("city")["total"].sum())

# Most sold product
print("\n Most Sold Product:")
print(df["name"].value_counts())

# Top customer
print("\n Top Customer:")
print(df.sort_values(by="total", ascending=False).head(1))
```
