### Dictionaries (Complete Training)
### STEP 1: What is a Dictionary?

👉 A dictionary stores data in key-value pairs

```python
student = {
    "name": "Ali",
    "marks": 80
}
```

### Think Like This:

| Key   | Value |
|-------|-------|
| name  | Ali   |
| marks | 80    |

👉 **Just like real-world:**
- `name` → label  
- `Ali` → actual data
###  STEP 2: Access Data

```python
print(student["name"])
```

### STEP 4: Update Data

```python
student["marks"] = 90
```

### STEP 5: Loop Through Dictionary

```python
for key, value in student.items():
    print(key, ":", value)
```

### Mini PROJECT
### Project Name:
👉 Student Record Manager (Python)
### Features:
- Store student data
- Add multiple students
- Display all students
- Calculate average marks

### FULL PROFESSIONAL CODE

```python
# Function to add a single student
def add_student():
    # Take student name input
    name = input("Enter student name: ")

    # Loop until valid marks are entered
    while True:
        try:
            # Convert input into float (numbers with decimals allowed)
            marks = float(input("Enter marks: "))
            break  # Exit loop if input is valid
        except ValueError:
            # Show error if input is not a number
            print("Please enter valid marks.")

    # Create a dictionary to store student data
    student = {
        "name": name,
        "marks": marks
    }

    # Return the student dictionary
    return student


# Function to display all students
def display_students(students):
    print("\n--- Student Records ---")

    # Loop through each student in the list
    for i, student in enumerate(students, start=1):
        print(f"\nStudent {i}:")

        # Loop through dictionary key-value pairs
        for key, value in student.items():
            print(f"{key.capitalize()}: {value}")


# Function to calculate average marks
def calculate_average(students):
    # If no students exist, return 0
    if len(students) == 0:
        return 0

    # Calculate total marks using list comprehension
    total = sum(student["marks"] for student in students)

    # Return average
    return total / len(students)


# Main function (program execution starts here)
def main():
    # Create empty list to store all students
    students = []

    print("Student Record Manager\n")

    # Loop to add multiple students
    while True:
        # Add student and store in list
        student = add_student()
        students.append(student)

        # Ask user if they want to continue
        choice = input("Add another student? (yes/no): ").lower()

        # Exit loop if user says no
        if choice != "yes":
            break

    # Display all student records
    display_students(students)

    # Calculate average marks
    avg = calculate_average(students)

    # Print average
    print(f"\n Average Marks: {avg}")


# Standard Python entry point
if __name__ == "__main__":
    main()
```
