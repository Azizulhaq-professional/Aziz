### STEP 1: LIST
👉 A list stores multiple values in one variable

```python
numbers = [10, 20, 30, 40, 50]
```

Why important professionally?

- Data science → datasets
- Cybersecurity → log files
- AI → training data
- Web apps → user records

### STEP 2: LOOP
👉 Loop = repeat action

Used when we know how many items exist.

```python
for num in numbers:
    print(num)
```

Professional use:

- Loop through files
- Loop through database rows
- Loop through API responses

### STEP 3: SUM LOGIC

👉 We use a variable to store total

```python
total = 0

for num in numbers:
    total += num

print("Sum:", total)
```

This is basic data processing.

### STEP 4 — WHILE LOOP

Used when we don’t know how many times loop will run.

Example: count until condition becomes false.

```python
count = 1

while count <= 5:
    print("Count:", count)
    count += 1
```

Professional use:

- Password attempts
- Waiting for user input
- Reading live data streams
### STEP 5 — DO-WHILE (Python Style)
Python has no real do-while loop.

But we simulate it using while True.

👉 Runs at least once.

```python
while True:
    num = int(input("Enter number > 10: "))

    if num > 10:
        print("Valid number")
        break
    else:
        print("Try again")
```

Professional use:

- Login systems
- Input validation
- Chatbots
### STEP 6 — IF / ELIF / ELSE

Decision making = core programming skill.

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 70:
    print("Grade B")
else:
    print("Grade C")
```

### STEP 7 — NESTED IF (Decision inside decision)

```python
age = 25
salary = 50000

if age >= 18:
    if salary >= 30000:
        print("Eligible for loan")
    else:
        print("Salary too low")
else:
    print("Under age")
```

Real world:

- Job filters
- Loan approval systems
- Security access systems

### STEP 8 — NESTED LOOP (Loop inside loop)

Used for tables, matrices, patterns, AI data.

```python
for i in range(3):
    for j in range(3):
        print(i, j)
```

Professional use:

- Image processing
- Machine learning
- Game development

### MINI PROJECT — Student Marks Analyzer 🎓

This combines everything 👇

```python
marks = [55, 70, 90, 40, 85]

total = 0
passed = 0
failed = 0

for mark in marks:
    total += mark

    if mark >= 50:
        passed += 1
    else:
        failed += 1

average = total / len(marks)

print("Average:", average)
print("Passed:", passed)
print("Failed:", failed)
```
