### Use Python open() and write data


```python
# Create CSV file in Google Colab

data = """name,product,price,quantity,city
Ali,Laptop,80000,1,Islamabad
Ahmed,Mobile,50000,2,Lahore
Sara,Tablet,30000,1,Karachi
Usman,Laptop,80000,1,Islamabad
Ayesha,Mobile,50000,3,Lahore"""

# Create and write file
with open("sales.csv", "w") as file:
    file.write(data)

print("CSV file created successfully!")
```

### FULL PROJECT CODE

```python
import pandas as pd

# Load dataset
df = pd.read_csv("sales_data.csv")

# Show data
print("Dataset Preview:")
print(df.head())

# Create total column
df["total"] = df["price"] * df["quantity"]

# Total revenue
total_revenue = df["total"].sum()
print("\nTotal Revenue:", total_revenue)

# Average price
avg_price = df["price"].mean()
print("Average Price:", avg_price)

# Sales by city
city_sales = df.groupby("city")["total"].sum()
print("\nSales by City:")
print(city_sales)
```
