### Student Management System
We will build a menu-driven program that stores students in a JSON file.
### STEP 1 — Import JSON

We need JSON to save data permanently.

```python
import json
```

### STEP 2 — Read JSON File Function

This function loads students from file.

```python
def read_json():
    try:
        with open("students.json", "r") as file:
            return json.load(file)
    except FileNotFoundError:
        return []
```

### Explanation
- Opens file in read mode
- json.load() converts JSON → Python list
- If file doesn't exist → return empty list
### STEP 3 — Save to JSON Function

```python
def save_to_json(students):
    with open("students.json", "w") as file:
        json.dump(students, file, indent=4)
```

### Explanation
- Saves updated student list to file
- indent=4 makes file readable
### STEP 4 — Add Student Function

```python
def add_student():
    name = input("Enter student name: ")
    marks = int(input("Enter marks: "))
    city = input("Enter city: ")

    student = {
        "name": name,
        "marks": marks,
        "city": city
    }

    students = read_json()
    students.append(student)
    save_to_json(students)

    print("Student added successfully!\n")
```

### Concepts used
- Dictionary
- List
- File handling
### STEP 5 — Display Students Function

```python
def display_students():
    students = read_json()

    if not students:
        print("No students found.\n")
        return

    print("\nStudent List")
    for s in students:
        print(f"Name: {s['name']} | Marks: {s['marks']} | City: {s['city']}")
    print()
```

### Concepts used
- Loop
- List of dictionaries
### STEP 6 — Calculate Average Marks

```python
def calculate_average():
    students = read_json()

    if not students:
        print("No data available.\n")
        return

    total = 0
    for s in students:
        total += s["marks"]

    avg = total / len(students)
    print(f"Average Marks: {avg:.2f}\n")
```

### STEP 7 — Search Student Function

```python
def search_student():
    name = input("Enter name to search: ").lower()
    students = read_json()

    found = False
    for s in students:
        if s["name"].lower() == name:
            print(f"Found: {s}")
            found = True

    if not found:
        print("Student not found.\n")
```

### STEP 8 — Main Menu (System Flow)

```python
def main():
    while True:
        print("1. Add Student")
        print("2. View Students")
        print("3. Search Student")
        print("4. Show Average")
        print("5. Exit")

        choice = input("Choose option: ")

        if choice == "1":
            add_student()
        elif choice == "2":
            display_students()
        elif choice == "3":
            search_student()
        elif choice == "4":
            calculate_average()
        elif choice == "5":
            print("Goodbye")
            break
        else:
            print("Invalid choice\n")
```

### STEP 9 — Run Program

```python
main()
```

### Student Management System — Full Code

```python
# ==============================
# Student Management System
# Concepts Used:
# Functions, Lists, Dictionaries,
# Loops, File Handling, JSON
# ==============================

import json   # JSON library to save data permanently


# ---------------------------------
# Function: Read data from JSON file
# ---------------------------------
def read_json():
    """
    This function reads students from students.json file.
    If file does not exist, it returns an empty list.
    """
    try:
        with open("students.json", "r") as file:
            return json.load(file)   # Convert JSON → Python list
    except FileNotFoundError:
        return []   # If file not found, start with empty list


# ---------------------------------
# Function: Save data into JSON file
# ---------------------------------
def save_to_json(students):
    """
    This function saves the student list into students.json
    """
    with open("students.json", "w") as file:
        json.dump(students, file, indent=4)  # Save nicely formatted JSON


# ---------------------------------
# Function: Add new student
# ---------------------------------
def add_student():
    """
    Take input from user and add new student
    """
    name = input("Enter student name: ")
    marks = int(input("Enter marks: "))
    city = input("Enter city: ")

    # Create dictionary for one student
    student = {
        "name": name,
        "marks": marks,
        "city": city
    }

    # Read existing students
    students = read_json()

    # Add new student into list
    students.append(student)

    # Save updated list back to file
    save_to_json(students)

    print("Student added successfully!\n")


# ---------------------------------
# Function: Display all students
# ---------------------------------
def display_students():
    """
    Print all students stored in JSON file
    """
    students = read_json()

    if not students:
        print("No students found.\n")
        return

    print("\nStudent List")
    for s in students:   # Loop through list of students
        print(f"Name: {s['name']} | Marks: {s['marks']} | City: {s['city']}")
    print()


# ---------------------------------
# Function: Calculate average marks
# ---------------------------------
def calculate_average():
    """
    Calculate and show average marks of students
    """
    students = read_json()

    if not students:
        print("No data available.\n")
        return

    total = 0

    # Loop to calculate sum of marks
    for s in students:
        total += s["marks"]

    # Calculate average
    avg = total / len(students)

    print(f"Average Marks: {avg:.2f}\n")


# ---------------------------------
# Function: Search student by name
# ---------------------------------
def search_student():
    """
    Search student using name
    """
    name = input("Enter name to search: ").lower()
    students = read_json()

    found = False

    # Loop to search student
    for s in students:
        if s["name"].lower() == name:
            print(f"Found: {s}")
            found = True

    if not found:
        print("Student not found.\n")


# ---------------------------------
# Main Menu (Program starts here)
# ---------------------------------
def main():
    """
    Menu-driven system
    Runs until user chooses Exit
    """
    while True:
        print("====== Student Management System ======")
        print("1. Add Student")
        print("2. View Students")
        print("3. Search Student")
        print("4. Show Average")
        print("5. Exit")

        choice = input("Choose option: ")

        if choice == "1":
            add_student()
        elif choice == "2":
            display_students()
        elif choice == "3":
            search_student()
        elif choice == "4":
            calculate_average()
        elif choice == "5":
            print("Goodbye")
            break
        else:
            print("Invalid choice. Try again.\n")


# Run the program
main()
```
