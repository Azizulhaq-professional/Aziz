
### Step 2: Update Flask App

📄 app.py
```
from flask import Flask, request, render_template
import joblib

app = Flask(__name__)

# Load model
model = joblib.load("final_pipeline_model.pkl")

@app.route('/')
def home():
    return render_template('index.html')

@app.route('/predict', methods=['POST'])
def predict():
    # Get data from form
    age = int(request.form['Age'])
    salary = int(request.form['Salary'])
    experience = int(request.form['Experience'])
    gender = int(request.form['Gender'])

    # Feature engineering (same as training)
    salary_per_exp = salary / experience

    features = [[age, salary, experience, gender, salary_per_exp]]

    prediction = model.predict(features)

    result = "Will Purchase" if prediction[0] == 1 else "Will Not Purchase"

    return render_template('index.html', prediction_text=result)

if __name__ == "__main__":
    app.run(debug=True)
```
### What Just Happened?

👉 Flow:
```
User fills form → Flask receives data → Model predicts → Result shown on page
```

### Real Application Behavior

- User doesn’t see ML code
- Just inputs data
- Gets instant prediction

### Important Detail

👉 You MUST apply same logic as training:
```
salary_per_exp = salary / experience
```
👉 Otherwise:
```
Wrong input → Wrong prediction
```
