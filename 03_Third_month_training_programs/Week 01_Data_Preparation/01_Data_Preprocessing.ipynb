### Step 1 — Create a Raw Dataset
Run this in a Colab cell:

```python
import pandas as pd
import numpy as np

# Create a dirty dataset (real-world style)
data = {
    "Name": ["Ali", "Sara", "Ahmed", "Zain", "Ayesha", None, "Usman"],
    "Age": [21, 25, np.nan, 30, 22, 28, None],
    "Salary": [50000, 60000, 55000, np.nan, 52000, 58000, 62000],
    "Department": ["IT", "HR", "IT", None, "Finance", "HR", "IT"]
}

df = pd.DataFrame(data)

# Save as CSV (optional but good practice)
df.to_csv("data.csv", index=False)

# Show dataset
print(df)
```

### STEP 2 — Understand Data

```python
print(df.info())
print(df.describe())
```

What you learn:

- Column data types
- Missing values
- Mean, min, max of numeric data

### STEP 3 — Check Missing Values

```python
print(df.isnull().sum())
```

### STEP 4 — Handle Missing Values
#### Option 1 (Remove rows)

```python
df_drop = df.dropna()
print(df_drop)
```

#### Option 2 (BEST PRACTICE → Fill values)

```python
# Fill numeric columns
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

# Fill categorical column
df['Department'] = df['Department'].fillna(df['Department'].mode()[0])
```

### STEP 5 — Remove Duplicates/

```python
df = df.drop_duplicates()
```

### STEP 6 — Fix Data Types

```python
df['Age'] = df['Age'].astype(int)
df['Salary'] = df['Salary'].astype(int)
```

### STEP 7 — Final Check

```python
print("After Cleaning Missing Values:")
print(df.isnull().sum())

print("\nFinal Clean Dataset:")
print(df)
```
