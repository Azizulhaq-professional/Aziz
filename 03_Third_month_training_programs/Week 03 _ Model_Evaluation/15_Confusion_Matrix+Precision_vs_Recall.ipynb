### 2. Confusion Matrix (Core Concept)

👉 It shows actual vs predicted results

# Structure:

|           | Predicted 0 | Predicted 1 |
|-----------|-------------|-------------|
| Actual 0  | TN          | FP          |
| Actual 1  | FN          | TP          |

### Meaning:
- TP (True Positive) → Correct YES
- TN (True Negative) → Correct NO
- FP (False Positive) → Wrong YES
- FN (False Negative) → Wrong NO

### 3. Code Example
```
from sklearn.metrics import confusion_matrix, precision_score, recall_score

# Assume y_test and y_pred already exist
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:\n", cm)

# Precision
precision = precision_score(y_test, y_pred)

# Recall
recall = recall_score(y_test, y_pred)

print("Precision:", precision)
print("Recall:", recall)
```

### 4. Understanding with Example

Suppose output:
```
[[50 10]
 [5  35]]
 ```
#### Breakdown:
- TN = 50
- FP = 10
- FN = 5
- TP = 35

### 5. Precision
Formula:

👉 Precision = TP / (TP + FP)

Meaning:
```
When model says YES → how often is it correct?
```
Example:
```
TP = 35  
FP = 10  
Precision = 35 / (35+10) = 0.77
```
Use when:
- False positives are costly

👉 Example:

- Spam detection
- Fraud alerts

### 6. Recall
Formula:

👉 Recall = TP / (TP + FN)

Meaning:
```
Out of all real YES → how many did model catch?
```
Example:
```
TP = 35  
FN = 5  
Recall = 35 / (35+5) = 0.87
```
Use when:
- Missing a case is dangerous

👉 Example:

- Disease detection
- Fraud detection

### 7. Precision vs Recall Tradeoff

👉 Increasing one often decreases the other
Example:

| Scenario           | Priority   |
|--------------------|------------|
| Spam filter        | Precision  |
| Cancer detection   | Recall     |

### Real-World Thinking

👉 Ask yourself:
```
What is more dangerous:
False Alarm OR Missing a real case?
```
### F1 Score (Balance)

👉 Combines precision + recall

Formula:
```
F1 = 2 * (Precision * Recall) / (Precision + Recall)
```

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, precision_score, recall_score

# Step 1: Define dataset
data = {
    'Age': [25, 30, 35, 40, 45, 50, 55, 60],
    'Salary': [50000, 60000, 80000, 100000, 120000, 140000, 200000, 300000],
    'Experience': [1, 3, 5, 7, 9, 11, 13, 15],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female']
}
df = pd.DataFrame(data)

# Target labels (Purchased outcome)
y = [0, 0, 0, 1, 1, 1, 1, 1]

# Step 2: Encode categorical variable (Gender)
encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
# Now Male=1, Female=0 (example)

# Step 3: Define features (X) and target (y)
X = df[['Age', 'Salary', 'Experience', 'Gender']]

# Step 4: Split into train/test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Step 5: Scale numeric features (helps convergence)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 6: Train Logistic Regression model
model = LogisticRegression(max_iter=200)
model.fit(X_train_scaled, y_train)

# Step 7: Make predictions
y_pred = model.predict(X_test_scaled)

# Step 8: Evaluate model
cm = confusion_matrix(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print("Confusion Matrix:\n", cm)
print("Precision:", precision)
print("Recall:", recall)
````

- Encoding Gender → converts "Male"/"Female" into numeric values so the model can process them.

- Scaling features → ensures Salary (large values) doesn’t dominate optimization, avoids convergence warnings.

- Train/test split → keeps evaluation fair by testing on unseen data.

- Evaluation metrics → confusion matrix, precision, recall give deeper insight than accuracy alone.

###  Updated Code with F1‑score

👉 Combines precision + recall

F1 = 2 * (Precision * Recall) / (Precision + Recall)

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

# Example dataset
data = {
    "Age": [22, 25, 40, 52, 46, 56, 23, 30, 40, 60],
    "Salary": [20000, 25000, 70000, 50000, 65000, 90000, 22000, 40000, 55000, 95000],
    "Experience": [1, 2, 20, 25, 11, 30, 1, 5, 10, 15],
    "Gender": ["Male", "Female", "Female", "Male", "Male", "Female", "Female", "Male", "Female", "Male"],
    "Purchased": [0, 0, 1, 1, 1, 1, 0, 0, 1, 1]
}
df = pd.DataFrame(data)

# Encode Gender column
encoder = LabelEncoder()
df["Gender"] = encoder.fit_transform(df["Gender"])
# Now Male=1, Female=0

# Features and target
X = df[["Age", "Salary", "Experience", "Gender"]]
y = df["Purchased"]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale numeric features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train model
model = LogisticRegression(max_iter=200)
model.fit(X_train_scaled, y_train)

# Predict
y_pred = model.predict(X_test_scaled)

# Evaluate
cm = confusion_matrix(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Confusion Matrix:\n", cm)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
````
