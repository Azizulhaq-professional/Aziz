### STEP 6: INDUSTRY TASK
Now we move from basic → professional thinking

##### 🔥 Build: User Info System (Version 1)
##### 🎯 Requirements:
- Take name
- Take age
- Take city
- Print formatted output

### Code

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))
city = input("Enter your city: ")

print("\n--- User Information ---")
print("Name:", name)
print("Age:", age)
print("City:", city)
```

### STEP 7: MINI CHALLENGE
