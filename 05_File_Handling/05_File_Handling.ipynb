### File Handling (Complete Training)
### STEP 1: What is File Handling?

👉 File handling = Read / Write data to files

Why important?
- Store user data
- Save logs
- Build databases
- Used in APIs

### STEP 2: File Modes

| Mode | Meaning |
|------|---------|
| "w"  | Write (overwrite file) |
| "a"  | Append (add data) |
| "r"  | Read file |

### STEP 3: Basic Example

```python
# write into file
file = open("data.txt", "w")
file.write("Hello World")
file.close()

# read file to show output
file = open("data.txt", "r")
content = file.read()
print(content)
file.close()
```

👉 Creates file and writes data

### Better Method (Professional)

```python
# Write to file
with open("data.txt", "w") as file:
    file.write("Hello World")

# Read from file and print
with open("data.txt", "r") as file:
    content = file.read()
    print(content)
```

#### Simple rule to remember
- "w" → write to file
- "r" → read from file
- print() → show on screen

### PROJECT
##### Project Name:

###### Student Record Saver (File System)

### Features:
- Take student data
- Save to file
- Read and display data

### FULL PROFESSIONAL CODE

```python
# Function to add student
def add_student():
    name = input("Enter student name: ")

    while True:
        try:
            marks = float(input("Enter marks: "))
            break
        except ValueError:
            print("Invalid input. Enter numbers only.")

    return {"name": name, "marks": marks}


# Function to save student to file
def save_to_file(student):
    # Open file in append mode (adds data without deleting old data)
    with open("students.txt", "a") as file:
        # Save data in simple format
        file.write(f"{student['name']},{student['marks']}\n")


# Function to read students from file
def read_from_file():
    print("\n--- Saved Student Records ---")

    try:
        with open("students.txt", "r") as file:
            for line in file:
                name, marks = line.strip().split(",")
                print(f"Name: {name}, Marks: {marks}")
    except FileNotFoundError:
        print("No records found.")


# Main function
def main():
    print("Student Record File System\n")

    while True:
        student = add_student()

        # Save data to file
        save_to_file(student)

        choice = input("Add another student? (yes/no): ").lower()
        if choice != "yes":
            break

    # Display saved data
    read_from_file()


if __name__ == "__main__":
    main()
```
