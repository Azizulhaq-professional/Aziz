### Goal

By the end, you will:

### Step 1: What is POST?
👉 Simple idea:

| Method | Purpose    |
|--------|------------|
| GET    | Get data   |
| POST   | Send data  |

**Example:**
- Login form → POST
- Signup → POST
- ML prediction → POST
### Step 2: Basic POST API Code
```
from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/submit', methods=['POST'])
def submit_data():
    data = request.json   # get JSON data
    
    name = data.get('name')
    age = data.get('age')
    
    return jsonify({
        "message": f"Hello {name}, you are {age} years old"
    })

if __name__ == '__main__':
    app.run(debug=True)
```

Inside folder:

Paste this code:

from flask import Flask, request, jsonify

# create flask app
app = Flask(__name__)

# create POST API
@app.route('/submit', methods=['POST'])
def submit_data():
    # get JSON data sent by client
    data = request.json

    # extract values from JSON
    name = data.get('name')
    age = data.get('age')

    # send JSON response back
    return jsonify({
        "message": f"Hello {name}, you are {age} years old"
    })

# run server
if __name__ == '__main__':
    app.run(debug=True)
```

````

### Step 4: Test Using Postman

Open Postman
- Setup:
- Method: POST
- URL:
```
http://127.0.0.1:5000/submit

```
#### Body → JSON:
```
{
  "name": "Aziz",
  "age": 23
}
```
### Response:
```{
  "message": "Hello Aziz, you are 23 years old"
}


### Step 4: How to Test API Using Postman
#### 1: Before Opening Postman (VERY IMPORTANT)

Make sure your API is running.

👉 In terminal:
```
python app.py
```
We should see:
```
Running on http://127.0.0.1:5000/
```
If this is not running → Postman will NOT work
####  2: Open Postman
- Launch Postman
- Click "New" → "HTTP Request"
####  3: Set Request Type

At the top left:

👉 Change method from GET → POST

(Click dropdown → select POST)
####  4: Enter URL

Paste this:
```
http://127.0.0.1:5000/submit
```
👉 This must match your Flask route:
```
@app.route('/submit', methods=['POST'])
```
####  5: Go to Body Section

This is the MOST IMPORTANT step.
- Click Body tab
- Select raw
- On the right side, change type to JSON

👉 (Dropdown usually says "Text" → change to "JSON")
####  6: Write JSON Data

Now paste this: 
```
{
  "name": "Aziz",
  "age": 23
}
```
Step 6: Click Send

👉 Press Send button
####  8: See Response

You will get:
```
{
  "message": "Hello Aziz, you are 23 years old"
}
```
#### What Just Happened (Simple Explanation)
- Postman sent data → API
- API received → request.json
- API processed → name + age
- API returned → response

### Step 5: What Happened? (Deep Understanding)
#### 1. Client Sends JSON

👉 Who is the client?
- It can be Postman, a website, or a mobile app.

👉 When you clicked Send, this happened:

You sent this data:
```
{
  "name": "Aziz",
  "age": 23
}
```
📦 This is called JSON (JavaScript Object Notation)

👉 It’s just a structured way to send data.
#### 2. API Receives Data (request.json)

Inside your Flask code:
```
data = request.json
```
👉 What this does:
- Takes incoming JSON data
- Converts it into a Python dictionary

So now:
```
data = {
  "name": "Aziz",
  "age": 23
}
```
👉 Then:
```
name = data.get('name')
age = data.get('age')
```
Now:
- name = "Aziz"
- age = 23
#### 3. Server Processes Data

👉 Server = your Flask app

Now your code runs logic:
```
"Hello Aziz, you are 23 years old"
```
💡 This is where you can:
- Do calculations
- Apply ML model
- Validate data
- Store in database

👉 Right now, you're just formatting a message.
#### 4. API Returns Response (jsonify())
```
return jsonify({
    "message": f"Hello {name}, you are {age} years old"
})
```
👉 What jsonify() does:
- Converts Python dictionary → JSON
- Sends it back to client
So response becomes:
```
{
  "message": "Hello Aziz, you are 23 years old"
}
```
### Full Flow
```
Postman (Client)
        ↓
Send JSON data
        ↓
Flask API (Server)
        ↓
request.json (receive data)
        ↓
Process data (Python logic)
        ↓
jsonify() (convert to JSON)
        ↓
Response sent back
        ↓
Postman shows result
```
### Real-Life Analogy (Easy Way to Remember)

Think of it like ordering food:

1. 🍔 You (client) → send order (JSON)
2. 👨‍🍳 Kitchen (server) → receives order (request.json)
3. 🍳 Cook prepares food (processing)
4. 🍽️ Food served back (jsonify() response)
### Why This is Important
This same flow is used in:
- AI apps → send input → get prediction
- Websites → login/signup
- Mobile apps → fetch user data
- Cybersecurity tools → send logs / analyze data
