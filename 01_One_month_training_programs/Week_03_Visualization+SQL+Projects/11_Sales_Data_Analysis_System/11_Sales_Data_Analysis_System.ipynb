```python
# ==========================================
# Sales Data Analysis System (COLAB VERSION)
# Author: Aziz Ul Haq
# Description:
# Complete end-to-end sales analysis using Pandas
# ==========================================

import pandas as pd

# ==================================================
# STEP 1 — UPLOAD CSV FILE (Required in Google Colab)
# ==================================================
from google.colab import files
uploaded = files.upload()   # Upload sales_data.csv here
```

```python
# ==================================================
# STEP 2 — LOAD DATA
# ==================================================
def load_data():
    """
    Load CSV file into DataFrame
    """
    try:
        df = pd.read_csv("sales.csv")
        print(" Data Loaded Successfully\n")
        return df
    except FileNotFoundError:
        print(" File not found! Upload sales_data.csv first.")
        return None


# ==================================================
# STEP 3 — CLEAN DATA
# ==================================================
def clean_data(df):
    """
    Clean dataset:
    - Handle missing values
    - Remove duplicates
    - Clean text
    - Create total column
    """

    print(" Cleaning Data...\n")

    # Fill missing values
    df["price"].fillna(df["price"].mean(), inplace=True)
    df["quantity"].fillna(1, inplace=True)

    # Remove duplicates
    df.drop_duplicates(inplace=True)

    # Remove empty names
    df.dropna(subset=["name"], inplace=True)

    # Clean text
    df["name"] = df["name"].str.strip()
    df["city"] = df["city"].str.strip()

    # Create total column
    df["total"] = df["price"] * df["quantity"]

    print(" Data Cleaned Successfully\n")
    return df


# ==================================================
# STEP 4 — SHOW SUMMARY
# ==================================================
def show_summary(df):
    print(" DATA SUMMARY")
    print(df.head())

    print("\n Columns:", df.columns.tolist())

    print("\n Data Info:")
    df.info()


# ==================================================
# STEP 5 — ANALYZE DATA
# ==================================================
def analyze_data(df):

    print("\n Total Revenue:", df["total"].sum())

    print("\n Sales by Product:")
    print(df.groupby("name")["total"].sum())

    print("\n Sales by City:")
    print(df.groupby("city")["total"].sum())

    print("\n Most Sold Products:")
    print(df["name"].value_counts())

    print("\n Highest Sale Record:")
    print(df.sort_values(by="total", ascending=False).head(1))


# ==================================================
# STEP 6 — FILTER DATA
# ==================================================
def filter_data(df):

    print("\n High Price Products (> 60000):")
    print(df[df["price"] > 60000])

    print("\n Islamabad Sales:")
    print(df[df["city"] == "Islamabad"])


# ==================================================
# MAIN FUNCTION (RUN COMPLETE SYSTEM)
# ==================================================
def main():

    df = load_data()

    if df is None:
        return

    df = clean_data(df)
    show_summary(df)
    analyze_data(df)
    filter_data(df)


# ==================================================
# RUN PROGRAM
# ==================================================
main()
```
