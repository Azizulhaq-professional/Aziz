### STEP 1 — Import Libraries

### TASK 1 — DATA CLEANING
Goal

We must prepare messy data before analysis.

What we will do:
1. Remove missing values
2. Remove duplicates
3. Create new column total = price × quantity

```python
# Load data
df = pd.read_csv("sales.csv")

print("Original Data:")
print(df.head())
```

### Cleaning

```python
# Remove rows with missing values
df.dropna(inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Create Total column
df["total"] = df["price"] * df["quantity"]

print("\nCleaned Data:")
print(df.head())
```

### Explanation
| Step               | Why                              |
|--------------------|----------------------------------|
| dropna()           | Missing values break analysis    |
| drop_duplicates()  | Prevent wrong revenue            |
| total column       | Needed for sales calculations    |

### TASK 2 — BASIC ANALYSIS
#### 1. Total Revenue

```python
total_revenue = df["total"].sum()
print(" Total Revenue =", total_revenue)
```

#### 2. Top Selling Product

```python
top_product = df.groupby("product")["quantity"].sum().idxmax()
print(" Top Selling Product =", top_product)
```

### 3. City With Highest Sales

```python
top_city = df.groupby("city")["total"].sum().idxmax()
print(" Highest Revenue City =", top_city)
```

### TASK 3 — ADVANCED THINKING
#### Q1 — Which customer spent the most?

```python
top_customer = df.groupby("name")["total"].sum().idxmax()
print(" Highest Spending Customer =", top_customer)
```

👉 We group by customer and sum spending.
#### Q2 — Least Popular Product

```python
least_product = df.groupby("product")["quantity"].sum().idxmin()
print(" Least Popular Product =", least_product)
```

👉 Lowest quantity sold = least popular.
#### Q3 — City With Lowest Revenue

```python
low_city = df.groupby("city")["total"].sum().idxmin()
print(" Lowest Revenue City =", low_city)
```

### TASK 4 — VISUALIZATION

Visuals make  project look professional.
### Bar Chart — Product Sales

```python
plt.figure(figsize=(8,5))
df.groupby("product")["total"].sum().plot(kind="bar")
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```

### Pie Chart — City Sales

```python
plt.figure(figsize=(6,6))
df.groupby("city")["total"].sum().plot(kind="pie", autopct='%1.1f%%')
plt.title("Sales by City")
plt.ylabel("")
plt.show()
```
