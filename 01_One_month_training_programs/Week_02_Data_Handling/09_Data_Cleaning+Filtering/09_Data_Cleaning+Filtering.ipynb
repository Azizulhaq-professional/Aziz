### STEP 5: Handle Missing Values
#### Option 1: Remove rows

````python
df = df.dropna()
````

#### Option 2: Fill values

````python
df["price"].fillna(df["price"].mean(), inplace=True)
df["quantity"].fillna(1, inplace=True)
````

👉 Replace missing data

###  STEP 6: Remove Duplicates

````python
df = df.drop_duplicates()
````

👉 Removes repeated rows

###  STEP 7: Clean Text (IMPORTANT)

````python
df["name"] = df["name"].str.strip()
df["city"] = df["city"].str.strip()
````

👉 Removes extra spaces

### STEP 8: Filtering Data (VERY IMPORTANT)
#### Filter by City

````python
isb_data = df[df["city"] == "Islamabad"]
print(isb_data)
````

#### Filter High Sales

````python
high_sales = df[df["price"] > 60000]
print(high_sales)
````

#### Multiple Conditions

````python
result = df[(df["city"] == "Lahore") & (df["price"] > 40000)]
print(result)
````

### FULL CLEANING PROJECT CODE

````python
import pandas as pd

# Load data
import pandas as pd
df = pd.read_csv("sales.csv")
print("Original Data:")
print(df)

# Show missing values
print("\n Missing Values:")
print(df.isnull().sum())

# Fill missing values
df["price"].fillna(df["price"].mean(), inplace=True)
df["quantity"].fillna(1, inplace=True)

# Remove duplicates
df = df.drop_duplicates()

# Remove empty names
df = df.dropna(subset=["name"])

# Clean text
df["name"] = df["name"].str.strip()
df["city"] = df["city"].str.strip()

print("\n Cleaned Data:")
print(df)

# Create total column
df["total"] = df["price"] * df["quantity"]

# Filter example
print("\n Islamabad Sales:")
print(df[df["city"] == "Islamabad"])

# High price products
print("\n High Price Products:")
print(df[df["price"] > 60000])
````
