### 5. Stemming vs Lemmatization
#### Stemming (Rough Cutting)
```
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
print(stemmer.stem("running"))
```
👉 Output:
```
run
```

````python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
print(stemmer.stem("running"))
````

#### Lemmatization (Meaning-based)
```
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("running", pos='v'))
```
👉 Output:
```
run
```
Difference:
| Feature  | Stemming      | Lemmatization |
| -------- | ------------- | ------------- |
| Accuracy | Low           | High          |
| Speed    | Fast          | Slower        |
| Meaning  | Not preserved | Preserved     |


````python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("running", pos='v'))
````

### 6. FULL PIPELINE EXAMPLE (IMPORTANT)

````python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

text = "AI is transforming the world!"

# Lowercase
text = text.lower()

# Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# Tokenize
tokens = word_tokenize(text)

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered = [word for word in tokens if word not in stop_words]

print(filtered)
````

### Original Sentences:

### Step-by-Step Code

````python
import nltk
import string
from nltk.tokenize import word_tokenize

nltk.download('punkt')

sentences = [
    "Machine Learning is AMAZING!!!",
    "NLP is used in Chatbots.",
    "Python, Java, and C++ are languages."
]

for text in sentences:
    # Step 1: Lowercase
    text = text.lower()

    # Step 2: Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Step 3: Tokenization
    tokens = word_tokenize(text)

    print(tokens)
````

### Output + Explanation
#### Sentence 1:
```
['machine', 'learning', 'is', 'amazing']
```
Explanation:

- "Machine" → machine (lowercase)
- "AMAZING!!!" → amazing (punctuation removed)
- Sentence split into words

#### Sentence 2:
```
['nlp', 'is', 'used', 'in', 'chatbots']
```
👉 Explanation:

- "Chatbots." → chatbots
- Clean tokens ready for processing

### Sentence 3:
```
['python', 'java', 'and', 'c', 'are', 'languages']
```
👉 Important Insight:

- "C++" → becomes "c"

👉 Because + is removed as punctuation

This shows:

👉 Cleaning can sometimes remove useful info

(advanced NLP handles this better)

### Task 3: Stemming vs Lemmatization

````python
from nltk.stem import PorterStemmer, WordNetLemmatizer
import nltk

nltk.download('wordnet')

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

words = ["playing", "better", "running"]

for word in words:
    stem = stemmer.stem(word)
    lemma = lemmatizer.lemmatize(word, pos='v')

    print(f"Word: {word}")
    print(f"Stemming: {stem}")
    print(f"Lemmatization: {lemma}")
    print("-----")
````

### Output + Explanation
#### Word: playing
```
Stemming: play
Lemmatization: play
```
👉 Both correct

👉 No big difference here

#### Word: better
```
Stemming: better
Lemmatization: better
```
👉 Important:

- Stemming  doesn’t change
- Lemmatization also needs correct POS (e.g., adjective)

👉 If we improve it:
```
lemmatizer.lemmatize("better", pos='a')
```
👉 Output: good

 Insight:

👉 Lemmatization understands meaning

#### Word: running
```
Stemming: run
Lemmatization: run
```
👉 Both give correct base form

### Final Comparison (Concept Clarity)
| Word    | Stemming | Lemmatization     | Meaning               |
| ------- | -------- | ----------------- | --------------------- |
| playing | play     | play              | Same                  |
| better  | better  | good  (with pos) | Lemmatization smarter |
| running | run      | run               | Same                  |



### Key Understanding (VERY IMPORTANT)

#### Stemming:
- Cuts words blindly
- Faster
- Less accurate

#### Lemmatization:
- Uses dictionary + grammar
- More accurate
- Keeps meaning

#### Real Insight (Industry Level)

👉 In real NLP systems:

- Basic models → Stemming
- Advanced systems → Lemmatization
