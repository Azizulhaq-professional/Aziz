##  1. Problem with TF-IDF & BoW

### 👉 These methods:

- Treat words **independently**
- Ignore **meaning**

###  Example:

| Words | Problem |
|-------|---------|
| "good" and "great" | → treated as different |
| "king" and "queen" | → no relationship |

###  Result:

> Machines don't understand context

---

##  2. What are Word Embeddings?

###  Word Embeddings =

> Convert words into **vectors** that capture **meaning**

###  Example:

| Word | Vector Representation |
|------|----------------------|
| king | [0.2, 0.8, 0.5] |
| queen | [0.21, 0.79, 0.52] |

###  Key Property:

> **Similar words → similar vectors** ✔

---

##  3. Famous Concept

###  One of the most famous relationships:
```
king - man + woman ≈ queen
```
## 4. Types of Word Embeddings

### Word2Vec (most important)
- CBOW (predict word from context)
- Skip-gram (predict context from word)

### Others (just awareness):
- GloVe
- FastText


## 5. Implement Word2Vec (Practical)

We’ll use Gensim

### Install (if needed):


````python
pip install gensim
````

````python
# Code
from gensim.models import Word2Vec

sentences = [
    ["ai", "is", "powerful"],
    ["machine", "learning", "is", "amazing"],
    ["ai", "is", "future"],
    ["deep", "learning", "is", "powerful"]
]

model = Word2Vec(sentences, vector_size=50, window=2, min_count=1)

# Check vector
print(model.wv['ai'])
````

### 6. Find Similar Words
```
print(model.wv.most_similar('ai'))
```
👉 Output example:
```
[('learning', 0.85), ('powerful', 0.80)]
```


````python
print(model.wv.most_similar('ai'))
````

## 7. Understanding Parameters

| Parameter | Meaning |
|-----------|---------|
| `vector_size` | Size of vector (dimensionality) |
| `window` | Context size (how many words to look at) |
| `min_count` | Ignore rare words (below this frequency) |

### Example Values:
```python
vector_size = 100   # 100-dimensional word vectors
window = 5          # Look at 5 words before and after
min_count = 2       # Ignore words appearing less than 2 times

### STEP 2 — Import Library

````python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')  # fix tokenizer issue
````

### Create Sentences (Add 10 sentences  Task 1)

We must give Word2Vec lots of sentences so it can learn word relationships.

````python
sentences = [
    "AI is transforming the world",
    "Machine learning is powerful",
    "Deep learning is the future",
    "I love learning new technologies",
    "Python is great for data science",
    "Neural networks learn patterns",
    "AI can solve complex problems",
    "Data science is very exciting",
    "Natural language processing is amazing",
    "Technology is evolving fast",

    #  Add 10 more sentences
    "Learning never stops in technology",
    "Artificial intelligence is the future",
    "Big data drives business decisions",
    "Python makes machine learning easy",
    "AI helps doctors detect diseases",
    "Students love practical learning",
    "Automation saves time and effort",
    "Data analysis reveals insights",
    "AI is a powerful tool",
    "Knowledge is power in modern world"
]
````

### STEP 4 Tokenize Sentences

Word2Vec needs list of word lists.

````python
tokenized_sentences = [word_tokenize(sentence.lower()) for sentence in sentences]

print(tokenized_sentences[:3])
````

### STEP 5 — Train Word2Vec Model

````python
model = Word2Vec(
    sentences=tokenized_sentences,
    vector_size=100,   # length of word vector
    window=5,          # context window (nearby words)
    min_count=1,       # keep words appearing ≥1 times
    workers=4
)
````

What training means

Model learns:

- Words appearing together = related meaning
- Context defines meaning

Example learned relation:
```
AI ↔ learning ↔ technology ↔ data
```
### TASK 2 — Find Similar Words to "learning"

````python
model.wv.most_similar('learning')
````

Meaning of this result

The model discovered:

Words often used near learning:

- machine learning
- deep learning
- students learning

So it learned semantic relationships

### TASK 3 — Print Vector of Word "powerful"

````python
vector = model.wv['powerful']
print(vector)
````
