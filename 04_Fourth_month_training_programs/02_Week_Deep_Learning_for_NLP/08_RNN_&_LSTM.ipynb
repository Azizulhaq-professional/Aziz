### 2. What is RNN (Recurrent Neural Network)?

рџ‘‰ RNN reads text word by word (sequence)

### Concept:
```
Word1 в†’ Word2 в†’ Word3 в†’ Word4
```
рџ‘‰ Each step remembers previous words

 Memory idea:
 ```
 h1 в†’ h2 в†’ h3 в†’ h4
 ```


### 3. Problem with RNN

рџ‘‰ It forgets long sentences

рџ‘‰ Called: Vanishing Gradient Problem

### 4. Solution: LSTM (Long Short-Term Memory)

рџ‘‰ LSTM is improved version of RNN

рџ‘‰ It remembers important information

### Think like:
- RNN = short memory
- LSTM = long memory

### 5. LSTM Intuition (Simple)

LSTM has:

- Forget gate вќЊ (remove useless info)
- Input gate вћ• (add new info)
- Output gate рџ“¤ (send info forward)

### 8. Compile & Train

#### Step 1 вЂ” Import libraries

````python
import numpy as np
import tensorflow as tf
````

#### Step 2 вЂ” Prepare data


````python
texts = [
    "I love AI",
    "AI is amazing",
    "I hate spam",
    "This is spam message"
]

labels = [1, 1, 0, 0]
````

#### Step 3 вЂ” Tokenization + Padding

````python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

seq = tokenizer.texts_to_sequences(texts)

# Make equal length sentences
X = pad_sequences(seq, padding='post')

print("Before conversion:", type(X))
````

### Step 4 вЂ”  Convert to Tensor (THE FIX)

````python
X = tf.convert_to_tensor(X)
y = tf.convert_to_tensor(labels)

print("After conversion:", type(X))
````

Now Keras understands the data

#### Step 5 вЂ” Build LSTM Model

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

vocab_size = len(tokenizer.word_index) + 1

model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=8, input_length=X.shape[1]),
    LSTM(16),
    Dense(1, activation='sigmoid')
])
````

### Step 6 вЂ” Compile model

````python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
````

### Step 7 вЂ” Train model

````python
model.fit(X, y, epochs=10)
````

### 9. Prediction

````python
test = ["I love machine learning"]

seq = tokenizer.texts_to_sequences(test)
pad = pad_sequences(seq, maxlen=X.shape[1])

pred = model.predict(pad)

print("Positive" if pred[0] > 0.5 else "Negative")
````

## FULL UPDATED CODE


### Step 1 вЂ” Import libraries

````python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
````

### Step 2 вЂ” Add MORE DATA (Very Important)

We now create 14 sentences instead of 4

````python
texts = [
    "I love AI",
    "AI is amazing",
    "I hate spam",
    "This is spam message",

    "Machine learning is powerful",
    "Deep learning is the future",
    "I enjoy studying AI",
    "Neural networks are fascinating",
    "Spam messages are annoying",
    "I received a fake offer",
    "Win money now",
    "Claim your free prize",
    "Limited time offer",
    "Congratulations you won lottery"
]

labels = [
    1,1,0,0,   # original
    1,1,1,1,   # good/positive
    0,0,0,0,0,0  # spam/negative
]
````

рџ‘‰ More sentences = model sees more patterns.

### Step 3 вЂ” Tokenization + Padding

````python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)
X = pad_sequences(sequences, padding='post')

# Convert to Tensor (important in new Keras)
X = tf.convert_to_tensor(X)
y = tf.convert_to_tensor(labels)

print("Shape of data:", X.shape)
````

### Step 4 вЂ” Build IMPROVED LSTM Model

Change 1: LSTM(16) в†’ LSTM(32)

````python
vocab_size = len(tokenizer.word_index) + 1

model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=16, input_length=X.shape[1]),
    LSTM(32),   # в­ђ Increased memory capacity
    Dense(1, activation='sigmoid')
])
````

### Why LSTM(32) helps

LSTM units = memory cells.
| Units | Meaning       |
| ----- | ------------- |
| 16    | Small memory  |
| 32    | More memory вњ” |
| 64+   | Large memory  |

More memory в†’ model remembers longer patterns.

### Step 5 вЂ” Compile Model

````python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
````

### Step 6 вЂ” Train Model

 Change 2: epochs 10 в†’ 20

````python
model.fit(X, y, epochs=20)
````

### Why more epochs help

Epoch = one full pass through dataset.
| Epochs | Meaning                |
| ------ | ---------------------- |
| 5      | barely learns          |
| 10     | starts learning        |
| 20     | learns patterns well вњ” |

### What improvements you should observe

After these upgrades:
| Change      | Effect           |
| ----------- | ---------------- |
| More data   | Less overfitting |
| Bigger LSTM | Better memory    |
| More epochs | Better training  |

Model becomes more confident and stable.
