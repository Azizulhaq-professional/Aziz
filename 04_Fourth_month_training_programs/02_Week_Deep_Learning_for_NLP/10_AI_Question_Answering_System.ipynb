### Full Project Code

````python
# Step 1: Import pipeline from transformers
# Pipeline = ready-to-use NLP model
from transformers import pipeline


# Step 2: Load Question Answering model
# This model is already trained on large datasets
qa_pipeline = pipeline("question-answering")


# Step 3: Define context (knowledge base)
# This is the text where model will search answers
context = """
Artificial Intelligence is transforming industries like healthcare,
education, and finance. It is also widely used in cybersecurity,
automation, and data analysis.
"""


# Step 4: Ask a question
question = "Which industries are being transformed?"


# Step 5: Get answer from model
result = qa_pipeline(
    question=question,
    context=context
)


# Step 6: Print result
print("Answer:", result['answer'])
print("Confidence Score:", result['score'])
````

### 5. What Happened Internally?

👉 Model:

- Reads full context
- Understands question
- Finds relevant part
- Returns answer

👉 This is powered by Transformer models like BERT

### Step 2: Add Multiple Contexts

👉 Real systems don’t use one paragraph—they use multiple sources.

````python
contexts = [
    """Artificial Intelligence is used in healthcare and finance.""",
    """Machine Learning helps systems learn from data.""",
    """Cybersecurity uses AI to detect threats."""
]

question = "Where is AI used?"

best_answer = ""
best_score = 0

for context in contexts:
    result = qa_pipeline(question=question, context=context)

    # Select best answer based on confidence score
    if result['score'] > best_score:
        best_score = result['score']
        best_answer = result['answer']

print("Best Answer:", best_answer)
print("Confidence:", best_score)
````

#### Explanation:

- 👉 Loop through all contexts
- 👉 Model checks each one
- 👉 Picks best answer

#### Key Insight:

👉 This is how:

- Search engines
- Chatbots
- AI assistants

👉 work internally

import streamlit as st
from transformers import pipeline

qa_pipeline = pipeline("question-answering")

# User input
context = st.text_area("Enter Context:")
question = st.text_input("Ask a Question:")

if st.button("Get Answer"):
    if context and question:
        result = qa_pipeline(question=question, context=context)

    else:

````python
import streamlit as st
from transformers import pipeline

qa_pipeline = pipeline("question-answering")

st.title(" AI Question Answering System")

# User input
context = st.text_area("Enter Context:")
question = st.text_input("Ask a Question:")

if st.button("Get Answer"):
    if context and question:
        result = qa_pipeline(question=question, context=context)

        st.write("### Answer:")
        st.write(result['answer'])

        st.write("Confidence:", result['score'])
    else:
        st.write("Please enter both context and question")
````
