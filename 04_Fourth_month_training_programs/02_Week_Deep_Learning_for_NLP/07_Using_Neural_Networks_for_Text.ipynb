### 3. Step-by-Step Implementation

WeвЂ™ll build a DL-based text classifier

### Step 1: Prepare Data

````python
# Step 1: Prepare Data

texts = [
    "I love AI",
    "AI is amazing",
    "I hate spam",
    "This is spam message"
]

labels = [1, 1, 0, 0]
````

### Step 2: Tokenization

````python
# Step 2: Tokenization

from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)
print(sequences)

````

### рџ‘‰ Words в†’ numbers

### Step 3: Padding

рџ‘‰ Neural networks need equal-length input

````python
# Step 3: Padding

from tensorflow.keras.preprocessing.sequence import pad_sequences

X = pad_sequences(sequences, padding='post')
print(X)
````

### 4. Build Deep Learning Model

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, Flatten

vocab_size = len(tokenizer.word_index) + 1

model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=8, input_length=X.shape[1]),
    Flatten(),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])
````

### 5. Compile Model

````python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
````

### 6. Train Model

### Step 1 вЂ” Import Libraries

````python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, Flatten
````

### Step 2 вЂ” Prepare Data

````python
texts = [
    "I love AI",
    "AI is amazing",
    "I hate spam",
    "This is spam message"
]

labels = [1, 1, 0, 0]

#  IMPORTANT FIX
labels = np.array(labels)   # convert list в†’ numpy array
````

Why?
Neural networks understand numbers в†’ not Python lists.

### Step 3 вЂ” Tokenization (Words в†’ Numbers)

````python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)
print("Sequences:", sequences)
````

Each word now has an ID.

### Step 4 вЂ” Padding (Make Equal Length)

Neural networks need same input length.

````python
X = pad_sequences(sequences, padding='post')

print("Padded Data:\n", X)
````

#### VERY IMPORTANT FIX (Core Fix)

Convert X to tensor (new TF requirement):
```
X = tf.convert_to_tensor(X)
labels = tf.convert_to_tensor(labels)
```
Now both inputs are tensors

### Step 5 вЂ” Build Deep Learning Model

````python
vocab_size = len(tokenizer.word_index) + 1

model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=8),
    Flatten(),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])
````

### What each layer does

| Layer     | Purpose                     |
| --------- | --------------------------- |
| Embedding | learns word meaning vectors |
| Flatten   | converts matrix в†’ vector    |
| Dense     | learns patterns             |
| Sigmoid   | outputs spam/not spam       |

### Step 6 вЂ” Compile Model

````python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
````

### Step 7 вЂ” Train Model

````python
model.fit(X, labels, epochs=10)
````

Model learns quickly because dataset is tiny.

### What You Learned

This bug taught you an important DL rule:

| Before         | After         |
| -------------- | ------------- |
| Python list    |  Not allowed |
| NumPy / Tensor |  Required    |

### Pro Insight


### 7. Prediction


````python
test = ["AI is powerful"]

seq = tokenizer.texts_to_sequences(test)
pad = pad_sequences(seq, maxlen=X.shape[1])

pred = model.predict(pad)

print("Positive" if pred[0] > 0.5 else "Negative")
````
