### 1. Problem with LSTM (Why Transformers Exist)

Even though LSTM is powerful, it has limitations:

-  Processes words one by one (slow)
-  Hard to handle long sentences
-  Limited context understanding

### 2. Solution: Transformers

👉 Transformers process all words at once (parallel)

👉 They use something called Attention

### 3. What is Attention? (CORE IDEA)

👉 Attention =
Model focuses on important words in a sentence

### Example:
```
"The movie was not good"
```
👉 Important word = "not"

👉 Model gives more weight to:
```
not → high importance
good → less importance (because of "not")
```

### 4. Self-Attention (Simplified)

👉 Each word looks at:

- All other words
- Decides which ones matter

#### Concept:
```
Word → checks → all words → assigns importance
```

### 5. Why Transformers Are Powerful
| Feature             | LSTM    | Transformer |
| ------------------- | ------- | ----------- |
| Speed               | Slow    | Fast        |
| Context             | Limited | Strong      |
| Parallel processing | ❌       | ✔         |

### Task 1 — Try Normal Inputs

#### Step 1 — Import libraries

````python
# Deep learning library
import tensorflow as tf
from tensorflow import keras

# Text processing tools
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
````

### Step 2 — Create training dataset (add many sentences)

We will add extra sentences as required in your tasks.

````python
# Sentences (0 = normal, 1 = spam / negative sentiment)
sentences = [
    "I love this movie",
    "This film was amazing",
    "I enjoyed the story",
    "This was a great experience",
    "I hate this movie",
    "This film was terrible",
    "Worst movie ever",
    "I did not like the story",

    # Added extra sentences (Task requirement)
    "AI is changing the world",
    "Machine learning is powerful",
    "Deep learning is the future",
    "I love artificial intelligence",
    "Spam messages are annoying",
    "I hate spam messages",
    "This product is very bad",
    "The service was awful",
    "The book is on the table",
    "It is raining outside"
]

labels = [
    0,0,0,0,   # positive
    1,1,1,1,   # negative/spam
    0,0,0,0,   # positive
    1,1,1,1,   # negative
    0,0        # neutral/normal
]
````

### Step 3 — Create and FIT tokenizer (VERY IMPORTANT)

This step was missing in your notebook

````python
# Create tokenizer object
tokenizer = Tokenizer(num_words=1000)

# Learn vocabulary from sentences
tokenizer.fit_on_texts(sentences)
````

 👉 Tokenizer now knows vocabulary of your dataset.

### Step 4 — Convert sentences → numbers

Deep learning cannot read text.

It only understands numbers.

````python
# Convert sentences into sequences of numbers
sequences = tokenizer.texts_to_sequences(sentences)

# Make all sequences same length (padding)
padded = pad_sequences(sequences, maxlen=10, padding='post')
````

### Step 5 — Build LSTM Model (updated tasks applied)

We apply your tasks:

-  LSTM(32)
-  epochs=20

````python
model = keras.Sequential([
    keras.layers.Embedding(input_dim=1000, output_dim=16, input_length=10),
    keras.layers.LSTM(32),  # Task: increased neurons
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
````

### Step 6 — Train model (epochs=20)

#### Fix = Convert labels → NumPy array

This is the missing step.

Add this BEFORE model.fit():
```
import numpy as np

labels = np.array(labels)
```

That’s it

````python
import numpy as np

# Convert labels list → numpy array (VERY IMPORTANT)
labels = np.array(labels)

# Train model
model.fit(padded, labels, epochs=20)
````

### Step 7 — Create classifier function (Prediction)

Now tokenizer exists, so error will disappear.

````python
def classifier(sentence):

    # Convert sentence → numbers using SAME tokenizer
    seq = tokenizer.texts_to_sequences([sentence])

    # Pad sequence to match training length
    padded_seq = pad_sequences(seq, maxlen=10, padding='post')

    # Predict probability
    prediction = model.predict(padded_seq)[0][0]

    # Convert probability → label
    if prediction > 0.5:
        print("Spam / Negative ", prediction)
    else:
        print("Normal / Positive ", prediction)
````

### BIG CONCEPT We JUST LEARNED

Your error taught you an important rule:

👉 In NLP inference, you MUST reuse the SAME tokenizer used during training.

### Why?

Because tokenizer stores vocabulary mapping:
```
love → 5
hate → 12
movie → 8
```
If tokenizer is missing → model cannot understand new text.
