### 5. Improved Chatbot Code (WITH ROLE + PERSONALITY)

````python
from transformers import pipeline

# Load model
chatbot = pipeline("text-generation", model="gpt2")

# Step 1: Define role + personality
system_prompt = """
You are an AI teacher.
Explain concepts in simple, clear, and beginner-friendly language.
Always give examples.
Keep answers short and helpful.
"""

chat_history = system_prompt

print(" Smart AI Chatbot Started (type 'exit' to stop)")

while True:
    user_input = input("\nYou: ")

    if user_input.lower() == "exit":
        break

    # Step 2: Add user input
    chat_history += f"\nUser: {user_input}\nBot:"

    # Step 3: Limit memory
    chat_history = chat_history[-1000:]

    # Step 4: Generate response
    response = chatbot(
        chat_history,
        max_length=150,
        temperature=0.7
    )

    # Step 5: Extract reply
    full_text = response[0]['generated_text']
    bot_reply = full_text.split("Bot:")[-1].strip()

    print("Bot:", bot_reply)

    # Step 6: Save response
    chat_history += f" {bot_reply}"
````
