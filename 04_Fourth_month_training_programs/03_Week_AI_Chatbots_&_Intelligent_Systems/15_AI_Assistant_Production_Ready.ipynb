### 3. Add Chat Memory (Persistent)

👉 Instead of losing chat, we store it

utils.py
```
# utils.py

def update_memory(history, user_input, bot_reply):
    history.append({"user": user_input, "bot": bot_reply})
    return history
```


### 4. Improve API with Memory + Error Handling
```
# api.py

from flask import Flask, request, jsonify
from transformers import pipeline
from utils import update_memory

app = Flask(__name__)

chatbot = pipeline("text-generation", model="gpt2")

chat_memory = []

roles = {
    "teacher": "You are an AI teacher.",
    "interview": "You are an interview coach.",
    "strict": "Answer in 2-3 lines only."
}

@app.route("/chat", methods=["POST"])
def chat():
    try:
        data = request.json
        
        role = data.get("role", "teacher")
        message = data.get("message", "")
        
        prompt = roles[role] + "\nUser: " + message + "\nBot:"
        
        response = chatbot(prompt, max_length=150)
        reply = response[0]['generated_text'].split("Bot:")[-1]
        
        # Save memory
        update_memory(chat_memory, message, reply)
        
        # Logging
        with open("logs.txt", "a") as f:
            f.write(f"User: {message}\nBot: {reply}\n\n")
        
        return jsonify({"response": reply})
    
    except Exception as e:
        return jsonify({"error": str(e)})
```

### 5. Improve Frontend (Show Chat History)
```
# app.py

import streamlit as st
import requests

st.title("🤖 AI Assistant")

if "history" not in st.session_state:
    st.session_state.history = []

role = st.selectbox("Role", ["teacher", "interview", "strict"])

message = st.text_input("Ask something:")

if st.button("Send"):
    response = requests.post(
        "http://127.0.0.1:5000/chat",
        json={"role": role, "message": message}
    )
    
    result = response.json()
    
    if "response" in result:
        st.session_state.history.append((message, result["response"]))

# Display chat history
for user, bot in st.session_state.history:
    st.write(f"👤 You: {user}")
    st.write(f"🤖 Bot: {bot}")
```
