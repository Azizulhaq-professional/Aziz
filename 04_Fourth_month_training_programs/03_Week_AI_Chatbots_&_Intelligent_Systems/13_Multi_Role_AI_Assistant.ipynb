### 4. Multi-Role Chatbot (Core Logic)

chatbot = pipeline("text-generation", model="gpt2")

# Select role

````python
from transformers import pipeline

chatbot = pipeline("text-generation", model="gpt2")

# Select role
role_choice = input("Choose role (teacher/interview/strict): ").lower()

system_prompt = roles.get(role_choice, roles["teacher"])

chat_history = system_prompt

while True:
    user_input = input("\nYou: ")

    if user_input == "exit":
        break

    chat_history += f"\nUser: {user_input}\nBot:"
    chat_history = chat_history[-1000:]

    response = chatbot(
        chat_history,
        max_length=150,
        temperature=0.7
    )

    reply = response[0]['generated_text'].split("Bot:")[-1]

    print("Bot:", reply)

    chat_history += reply
````

### 6. Convert into API (IMPORTANT)

👉 Now we make it usable for:
- Web apps
- Mobile apps
- Frontend


Flask API
```
from flask import Flask, request, jsonify
from transformers import pipeline

app = Flask(__name__)

chatbot = pipeline("text-generation", model="gpt2")

roles = {
    "teacher": "You are an AI teacher. Explain simply.",
    "interview": "You are an interview coach.",
    "strict": "Answer in 2-3 lines only."
}

@app.route("/chat", methods=["POST"])
def chat():
    data = request.json
    
    role = data.get("role", "teacher")
    message = data.get("message", "")
    
    prompt = roles[role] + "\nUser: " + message + "\nBot:"
    
    response = chatbot(prompt, max_length=150)
    reply = response[0]['generated_text'].split("Bot:")[-1]
    
    return jsonify({"response": reply})

app.run()
```
