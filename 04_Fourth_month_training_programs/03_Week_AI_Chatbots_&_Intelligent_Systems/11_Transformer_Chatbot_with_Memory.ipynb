### Improved Chatbot Code (With All Enhancements)

````python
# Step 1: Import pipeline
from transformers import pipeline

# Step 2: Load model (GPT-style text generator)
chatbot = pipeline("text-generation", model="gpt2")

# Step 3: Initialize memory
chat_history = ""

print(" Chatbot started (type 'exit' to stop)")

while True:
    # Step 4: User input
    user_input = input("\nYou: ")

    if user_input.lower() == "exit":
        break

    # Step 5: Add user input to memory
    chat_history += f"User: {user_input}\nBot:"

    #  IMPROVEMENT 1: Limit memory size
    # Keep only last 1000 characters
    chat_history = chat_history[-1000:]

    # Step 6: Generate response with improved settings
    response = chatbot(
        chat_history,
        max_length=150,        #  IMPROVEMENT 2: Control response length
        temperature=0.7,       #  IMPROVEMENT 3: Improve quality
        num_return_sequences=1
    )

    # Step 7: Extract bot reply
    full_text = response[0]['generated_text']
    bot_reply = full_text.split("Bot:")[-1].strip()

    # Step 8: Print response
    print("Bot:", bot_reply)

    # Step 9: Save bot reply in memory
    chat_history += f" {bot_reply}\n"
````

### Now Understand Each Improvement (VERY IMPORTANT)

#### 1. Memory Limiting
```
chat_history = chat_history[-1000:]
```

#### Problem Without This:
- Chat keeps growing
- Input becomes too long
- Model becomes slow
- May crash

What This Does:

👉 Keeps only last 1000 characters

#### Real Meaning:
- Keeps recent conversation
- Removes old irrelevant chat

#### Analogy:

👉 Like human memory:

- You remember recent discussion better
- Forget old details

### 2. Control Response Length
```
max_length=150
```

#### Problem Without This:
- Model may generate:
  - Very long responses
  - Irrelevant text
  - Repetition

#### What This Does:

👉 Limits total output length

#### Real Meaning:
- Keeps response short and focused

#### Example:

Without limit:
```
Bot: AI is great... (continues for 10 lines )
```

With limit:
```
Bot: AI is great for solving problems
```

### 3. Temperature (MOST IMPORTANT)
```
temperature=0.7
```
### What is Temperature?

👉 Controls creativity vs accuracy

#### Values Meaning:
| Value | Behavior             |
| ----- | -------------------- |
| 0.2   | Very strict, robotic |
| 0.7   | Balanced (best) ✔    |
| 1.0+  | Creative but random  |

### What Happens Internally:

👉 Model chooses next word based on probability
- Low temperature → picks highest probability
- High temperature → explores more options

#### Example:

Input:
```
AI is
```
- Temp = 0.2 → "useful" (safe)
- Temp = 0.7 → "transforming industries"
- Temp = 1.2 → "flying bananas in space"

### Combined Effect (VERY IMPORTANT)
| Feature      | Result            |
| ------------ | ----------------- |
| Memory limit | Faster + relevant |
| Max length   | Controlled output |
| Temperature  | Better quality    |
