### 4. Backend (Flask API)
```
# api.py

from flask import Flask, request, jsonify
from transformers import pipeline

app = Flask(__name__)

chatbot = pipeline("text-generation", model="gpt2")

roles = {
    "teacher": "You are an AI teacher. Explain simply.",
    "interview": "You are an interview coach.",
    "strict": "Answer in 2-3 lines only."
}

@app.route("/chat", methods=["POST"])
def chat():
    data = request.json
    
    role = data.get("role", "teacher")
    message = data.get("message", "")
    
    prompt = roles[role] + "\nUser: " + message + "\nBot:"
    
    response = chatbot(prompt, max_length=150, temperature=0.7)
    reply = response[0]['generated_text'].split("Bot:")[-1]
    
    return jsonify({"response": reply})

if __name__ == "__main__":
    app.run(debug=True)
```

### 5. Frontend (Streamlit UI)
```
# app.py

import streamlit as st
import requests

st.title("🤖 AI Assistant")

# Role selection
role = st.selectbox(
    "Choose Role",
    ["teacher", "interview", "strict"]
)

# User input
message = st.text_input("Ask something:")

if st.button("Send"):
    if message:
        response = requests.post(
            "http://127.0.0.1:5000/chat",
            json={"role": role, "message": message}
        )
        
        result = response.json()
        
        st.write("### 🤖 Response:")
        st.write(result["response"])
    else:
        st.write("Please enter a message")
```
