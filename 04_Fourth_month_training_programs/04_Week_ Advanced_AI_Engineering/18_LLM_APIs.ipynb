## STEP 2 — Load an Instruction LLM (Important Upgrade)

```python
from transformers import pipeline

# Use "text-generation" instead of "text2text-generation"
llm = pipeline(
    "text-generation",
    model="google/flan-t5-base"
)

print("Instruction LLM ready")
```

## STEP 3 — Create Roles (System Prompts)

This is prompt engineering.

```python
roles = {
    "teacher": "Explain clearly in simple words with examples.",
    "interview": "Answer like a professional job candidate.",
    "strict": "Reply in only 2 short lines."
}
```

## STEP 4 — Build LLM API Function (OpenAI Replacement)

This function behaves like:

openai.ChatCompletion.create()

```python
def llm_api(role, message, history=[]):

    # Build conversation context
    context = ""
    for h in history:
        context += f"User: {h[0]} Bot: {h[1]}\n"

    prompt = f"""
    {roles[role]}
    Conversation:
    {context}
    User: {message}
    Bot:
    """

    output = llm(
        prompt,
        max_length=150,   # token control (cost control concept)
        temperature=0.7
    )

    reply = output[0]["generated_text"]
    return reply
```

This is now your FREE LLM API

## STEP 5 — Add Chat Memory (Conversation History)

```python
chat_memory = []
```

```python
# Test conversation:

reply = llm_api("teacher", "What is machine learning?", chat_memory)
print(reply)

chat_memory.append(("What is machine learning?", reply))
```

```python
# Continue conversation:

reply = llm_api("teacher", "Give a real life example", chat_memory)
print(reply)

chat_memory.append(("Give a real life example", reply))
```

You now have conversation memory like ChatGPT.

## STEP 7 — Cost & Performance Control (Key Concept)

Even though this is free, we simulate real API cost control.

```python
# Control response length

max_length=120
```

```python
# Control creativity

temperature=0.3  # factual
temperature=0.9  # creative
```

```python
# Modify function:

def llm_api(role, message, history=[], max_tokens=120, temp=0.7):
    context = ""
    for h in history:
        context += f"User: {h[0]} Bot: {h[1]}\n"

    prompt = f"{roles[role]}\n{context}User: {message}\nBot:"

    output = llm(prompt, max_length=max_tokens, temperature=temp)
    return output[0]["generated_text"]
```

This mimics production token control.

## STEP 8 — Simulate Production Backend Endpoint

This replaces OpenAI backend call.

```python
def chat_endpoint(user_id, role, message, user_sessions):

    if user_id not in user_sessions:
        user_sessions[user_id] = []

    history = user_sessions[user_id]
    reply = llm_api(role, message, history)

    history.append((message, reply))
    return reply
```

```python
# Create users:

import uuid
sessions = {}

user1 = str(uuid.uuid4())
user2 = str(uuid.uuid4())
```

```python
# Test:

print(chat_endpoint(user1, "interview", "Tell me about yourself", sessions))
print(chat_endpoint(user1, "interview", "What are your strengths?", sessions))
```

```python
# User2 separate:

print(chat_endpoint(user2, "teacher", "Explain Python", sessions))
```

Multi-user LLM system achieved
